```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier, VotingClassifier # Added VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay, f1_score # Added f1_score
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE, ADASYN # Added ADASYN
from collections import Counter
from imblearn.pipeline import Pipeline
import xgboost as xgb # Added
import lightgbm as lgb # Added
from scipy.stats import randint, uniform # For RandomizedSearchCV

# --- 1. โหลดข้อมูล ---
df = pd.read_csv('food_menu.csv') # โหลดไฟล์ CSV ที่เก็บข้อมูลเมนูอาหาร

# Make a copy to avoid SettingWithCopyWarning later if we modify df directly
df_processed = df.copy()

# --- [3] Feature Engineering ---
# df_processed['category_encoded'] = LabelEncoder().fit_transform(df_processed['category'])
# df_processed['suitable_for_disease'] = LabelEncoder().fit_transform(df_processed['suitable_for_disease'])
# Extract numeric from 'amount'
df_processed['amount_numeric'] = df_processed['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
df_processed['sugar_to_carb_ratio'] = df_processed['sugar'] / (df_processed['carbohydrate'] + 1e-8)
df_processed['fat_to_calorie_ratio'] = df_processed['fat'] / (df_processed['calories'] + 1e-8)
df_processed['sodium_per_calorie'] = df_processed['sodium'] / (df_processed['calories'] + 1e-8)


# Density & per-calorie features
df_processed['calorie_density'] = df_processed['calories'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['sugar_per_calorie'] = df_processed['sugar'] / (df_processed['calories'] + 1e-8)
df_processed['fat_per_calorie'] = df_processed['fat'] / (df_processed['calories'] + 1e-8)
df_processed['protein_per_calorie'] = df_processed['protein'] / (df_processed['calories'] + 1e-8)
df_processed['carb_per_calorie'] = df_processed['carbohydrate'] / (df_processed['calories'] + 1e-8)
df_processed['protein_to_fat_ratio'] = df_processed['protein'] / (df_processed['fat'] + 1e-8)

# Density by amount
df_processed['protein_density'] = df_processed['protein'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['fat_density'] = df_processed['fat'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['carb_density'] = df_processed['carbohydrate'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['sugar_density'] = df_processed['sugar'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['sodium_density'] = df_processed['sodium'] / (df_processed['amount_numeric'] + 1e-8)

# Extra ratios and densities
df_processed['fiber_per_calorie'] = df_processed['fiber'] / (df_processed['calories'] + 1e-8)
df_processed['fiber_to_carb_ratio'] = df_processed['fiber'] / (df_processed['carbohydrate'] + 1e-8)
df_processed['sodium_to_sugar_ratio'] = df_processed['sodium'] / (df_processed['sugar'] + 1e-8)
df_processed['saturated_fat_per_calorie'] = df_processed['saturated_fat'] / (df_processed['calories'] + 1e-8)
df_processed['saturated_fat_to_total_fat_ratio'] = df_processed['saturated_fat'] / (df_processed['fat'] + 1e-8)
df_processed['fiber_density'] = df_processed['fiber'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['saturated_fat_density'] = df_processed['saturated_fat'] / (df_processed['amount_numeric'] + 1e-8)

# --- NEW FEATURE ENGINEERING FOR LOW SUGAR ---
sugar_threshold = df_processed['sugar'].quantile(0.15) # Example: bottom 15%
sugar_per_cal_threshold = df_processed['sugar_per_calorie'].quantile(0.15) # Example: bottom 15%

df_processed['is_very_low_sugar'] = (df_processed['sugar'] <= sugar_threshold).astype(int)
df_processed['is_very_low_sugar_per_cal'] = (df_processed['sugar_per_calorie'] <= sugar_per_cal_threshold).astype(int)


# --- One-Hot Encode 'category' ---
# Drop the original 'category' column as we'll replace it with one-hot encoded ones
df_processed = pd.get_dummies(df_processed, columns=['category'], prefix='cat', drop_first=False) # drop_first=False to keep all, can be True if no multicollinearity issues arise

# --- [4] Select Features & Target ---
# Dynamically build feature_cols to include new one-hot encoded categories
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie', 'saturated_fat_to_total_fat_ratio',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density',
    'is_very_low_sugar', 'is_very_low_sugar_per_cal' # Add new binary features
]

# Add one-hot encoded category columns
category_ohe_cols = [col for col in df_processed.columns if col.startswith('cat_')]
feature_cols.extend(category_ohe_cols)

X = df_processed[feature_cols]
y = df_processed['food_type']

print("New X dtypes:")
print(X.dtypes)

# --- 5. Encode target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y) # แปลง label เป็นตัวเลข
print("\nFood Type Mapping:")
for idx, class_label in enumerate(le.classes_):
    print(f"{class_label} --> {idx}") # แสดง mapping ระหว่างชื่อ class กับตัวเลข

class_labels = le.classes_ # เก็บชื่อ class ไว้ใช้แสดงผลรายงานและกราฟ

# --- 6. Train/Test Split ---
# แบ่งข้อมูลเป็นชุดฝึกสอนและทดสอบ โดย stratify เพื่อให้แต่ละ class มีสัดส่วนเท่าเดิมในทั้งสองชุด
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)

# --- Define the Pipeline for Grid Search ---
# สร้าง pipeline ที่ประกอบด้วยขั้นตอนการจัดการข้อมูล (imputer, scaler, smote) และตัวจำแนก (classifier)
pipeline_steps = [
    ('imputer', SimpleImputer(strategy='mean')), # เติมค่าที่ขาดด้วยค่าเฉลี่ย
    ('scaler', StandardScaler()), # ปรับข้อมูลให้มีค่าเฉลี่ย 0 และส่วนเบี่ยงเบนมาตรฐาน 1
    ('smote', SMOTE(random_state=42, k_neighbors=3)), # เพิ่มข้อมูล class ที่น้อยด้วย SMOTE, adjusted k_neighbors
    ('classifier', None) # ที่ว่างสำหรับใส่โมเดลในขั้นตอน grid search
]


# --- 7. จัดการ missing values & 8. Scaling & 8.1 SMOTE (สำหรับ train/test evaluation โดยตรง) ---
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# Adjusted k_neighbors for SMOTE here too
smote = SMOTE(random_state=42, k_neighbors=3)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

print(f"\nBefore SMOTE: {Counter(y_train)}") # แสดงจำนวนตัวอย่างก่อนทำ SMOTE
print(f"After SMOTE: {Counter(y_train_balanced)}") # แสดงจำนวนตัวอย่างหลังทำ SMOTE
print(f"Training samples (after SMOTE): {X_train_balanced.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")


# --- 9. สร้างโมเดลและกำหนด Hyperparameter Grids สำหรับ Grid Search ---
param_grids = {
    "Random Forest": {
        'classifier': [RandomForestClassifier(random_state=42, class_weight='balanced')], # Add class_weight
        'classifier__n_estimators': [100, 200, 300], # Increased max
        'classifier__max_depth': [5, 10, 15, None], # Added 15
        'classifier__min_samples_split': [2, 5, 10], # Added 10
        'classifier__min_samples_leaf': [1, 2, 4],
        'classifier__max_features': ['sqrt', 'log2', None] # Added max_features
    },
    "KNN": {
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [3, 5, 7, 9, 11, 13, 15], # Wider range
        'classifier__weights': ['uniform', 'distance'],
        'classifier__metric': ['euclidean', 'manhattan', 'minkowski'] # Added minkowski
    },
    "Decision Tree": {
        'classifier': [DecisionTreeClassifier(random_state=42, class_weight='balanced')], # Ensure class_weight
        'classifier__max_depth': [5, 10, 15, None], # Added 15
        'classifier__min_samples_split': [2, 5, 10], # Added 10
        'classifier__min_samples_leaf': [1, 2, 4], # Added 4
        'classifier__max_features': ['sqrt', 'log2', None] # Added max_features
    },
    "XGBoost": { # Added XGBoost
        'classifier': [xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss', n_jobs=-1)],
        'classifier__n_estimators': [100, 200, 300],
        'classifier__max_depth': [3, 5, 7],
        'classifier__learning_rate': [0.01, 0.05, 0.1, 0.2],
        'classifier__subsample': [0.7, 0.8, 0.9, 1.0],
        'classifier__colsample_bytree': [0.7, 0.8, 0.9, 1.0]
    },
    "LightGBM": { # Added LightGBM
        'classifier': [lgb.LGBMClassifier(random_state=42, n_jobs=-1)],
        'classifier__n_estimators': [100, 200, 300],
        'classifier__max_depth': [5, 10, 15, -1], # -1 means no limit
        'classifier__learning_rate': [0.01, 0.05, 0.1, 0.2],
        'classifier__num_leaves': [20, 31, 50, 70],
        'classifier__min_child_samples': [20, 30, 40]
    }
}

best_models = {}
cv_results = {}
kfold_grid = StratifiedKFold(n_splits=5, shuffle=True, random_state=42) # กำหนด cross-validation แบบ stratified 5 fold

print("\n--- Performing Grid Search Cross-Validation ---")
for name, param_grid in param_grids.items():
    print(f"\nRunning Grid Search for {name}...")
    
    # สร้าง pipeline สำหรับ GridSearchCV โดยใส่โมเดลตัวแรกของแต่ละชนิดเป็นค่าเริ่มต้น
    current_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler()),
        # Consider adjusting k_neighbors for SMOTE here if specific class is too small
        ('smote', SMOTE(random_state=42, k_neighbors=3)), 
        ('classifier', param_grid['classifier'][0]) 
    ])
    
    # เอา parameter ยกเว้น 'classifier' ออกมาใช้กับ grid search
    grid_params = {k: v for k, v in param_grid.items() if k != 'classifier'}
    
    grid_search = GridSearchCV(
        estimator=current_pipeline,
        param_grid=grid_params,
        cv=kfold_grid,
        scoring='f1_weighted', # <--- Changed scoring to f1_weighted
        n_jobs=-1, # ใช้ core ทั้งหมดในเครื่อง
        verbose=1
    )
    
    # เรียกใช้ Grid Search 
    grid_search.fit(X_train, y_train)
    
    best_models[name] = grid_search.best_estimator_ # เก็บ pipeline ที่ดีที่สุด
    cv_results[name] = {
        'best_score': grid_search.best_score_,
        'best_params': grid_search.best_params_
    }
    
    print(f"Finished Grid Search for {name}.")
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best cross-validation F1-weighted: {grid_search.best_score_:.4f}") # Changed print message

# --- 10.1 ประเมินผลบน Train Set (using models from Grid Search) ---
train_accuracies = []
train_f1_scores = [] # Store F1 scores too
print("\n--- Evaluating Best Models (from Grid Search) on Train Set ---")
for name, model_pipeline in best_models.items():
    
    # Prediction on training data through the pipeline
    y_train_pred_pipeline = model_pipeline.predict(X_train)
    acc_train = accuracy_score(y_train, y_train_pred_pipeline)
    f1_train = f1_score(y_train, y_train_pred_pipeline, average='weighted')
    report_train = classification_report(y_train, y_train_pred_pipeline, target_names=class_labels)
    train_accuracies.append(acc_train)
    train_f1_scores.append(f1_train)

    print(f"\n--- {name} Train Results (Best Model - Pipeline Evaluation) ---")
    print(f"Accuracy: {acc_train:.2f}")
    print(f"F1-weighted: {f1_train:.2f}") # Added F1-weighted
    print("Classification Report:")
    print(report_train)

    # Display confusion matrix for train set (using pipeline's prediction on original X_train)
    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_pipeline, display_labels=class_labels,
                                             cmap='Blues', ax=ax)
    ax.set_title(f"Confusion Matrix: {name} (Train Set - Best Model Pipeline)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- 10.2 ประเมินผลบน Test Set (using models from Grid Search) ---
test_accuracies = []
test_f1_scores = [] # Store F1 scores too
model_names = []
print("\n--- Evaluating Best Models (from Grid Search) on Test Set ---")
for name, model_pipeline in best_models.items():
    # pipeline จะทำ preprocessing ให้เอง (imputation, scaling) ก่อนทำนาย แต่ SMOTE จะไม่ถูกใช้ตอน predict (ถูกต้องแล้ว)
    y_test_pred = model_pipeline.predict(X_test)
    acc = accuracy_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred, average='weighted') # Calculate F1-weighted
    report = classification_report(y_test, y_test_pred, target_names=class_labels)

    test_accuracies.append(acc)
    test_f1_scores.append(f1) # Store F1
    model_names.append(name)

    print(f"\n--- {name} Test Results (Best Model) ---")
    print(f"Accuracy: {acc:.2f}")
    print(f"F1-weighted: {f1:.2f}") # Added F1-weighted
    print("Classification Report:")
    print(report)
    print(f"Best parameters: {cv_results[name]['best_params']}")

    # แสดง confusion matrix สำหรับ test set
    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels,
                                             cmap='Blues', ax=ax)
    ax.set_title(f"Confusion Matrix: {name} (Test Set - Best Model)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- 10.3 แสดงผล Cross Validation ที่ได้จาก Grid Search ---
print("\n--- Cross-Validation Results from Grid Search ---")
for name, gs_result in cv_results.items():
    print(f"{name}: Best CV F1-weighted = {gs_result['best_score']:.3f}, Best Params = {gs_result['best_params']}")

# --- 11.1 กราฟเปรียบเทียบ Accuracy บน Train Set ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=train_accuracies, y=list(best_models.keys()), palette="viridis", orient='h')
for i, v in enumerate(train_accuracies):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Accuracy Comparison on Train Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 11.2 กราฟเปรียบเทียบ F1-weighted บน Train Set (NEW PLOT) ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=train_f1_scores, y=list(best_models.keys()), palette="viridis", orient='h')
for i, v in enumerate(train_f1_scores):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model F1-weighted Comparison on Train Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('F1-weighted Score')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 11.3 กราฟเปรียบเทียบ Accuracy บน Test Set ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=test_accuracies, y=model_names, palette="magma", orient='h')
for i, v in enumerate(test_accuracies):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Accuracy Comparison on Test Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 11.4 กราฟเปรียบเทียบ F1-weighted บน Test Set (NEW PLOT) ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=test_f1_scores, y=model_names, palette="magma", orient='h')
for i, v in enumerate(test_f1_scores):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model F1-weighted Comparison on Test Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('F1-weighted Score')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()


# --- 11.5 กราฟเปรียบเทียบ Cross Validation F1-weighted จาก Grid Search (Updated Plot) ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=[res['best_score'] for res in cv_results.values()], y=list(cv_results.keys()), palette="coolwarm", orient='h')
for i, (name, res) in enumerate(cv_results.items()):
    ax.text(res['best_score'] + 0.01, i, f"{res['best_score']:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Best Cross-Validated F1-weighted Score (from Grid Search)', fontsize=16)
plt.xlabel('F1-weighted Score')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 12. แสดง Feature Importance สำหรับโมเดลที่รองรับ (Random Forest, Decision Tree, XGBoost, LightGBM) ---
feature_names = X.columns.tolist()

for name, model_pipeline in best_models.items():
    classifier = model_pipeline.named_steps['classifier']
    
    if hasattr(classifier, 'feature_importances_'):
        importances = classifier.feature_importances_
        indices = np.argsort(importances)[::-1]  # เรียงลำดับจากความสำคัญมากไปน้อย

        plt.figure(figsize=(10, max(6, len(feature_names) * 0.3))) # Adjust figure height based on number of features
        sns.barplot(x=importances[indices], y=np.array(feature_names)[indices], palette='crest')
        plt.title(f"Feature Importance for {name} (Best Model)")
        plt.xlabel('Importance')
        plt.ylabel('Feature')
        plt.tight_layout()
        plt.show()
    else:
        print(f"\n{name} ไม่รองรับการแสดง Feature Importance (เช่น KNN).")

# --- 13. ลองใช้ VotingClassifier (Optional, after good individual models) ---
print("\n--- Attempting Voting Classifier ---")
try:
    # Prepare estimators for VotingClassifier: (name, trained_pipeline)
    # The pipelines themselves are the estimators here.
    estimators_for_voting = [(name, model_pipeline) for name, model_pipeline in best_models.items() 
                             if name not in ["KNN"]] # KNN doesn't output probabilities easily for 'soft' voting

    if estimators_for_voting:
        # Use 'soft' voting if classifiers support predict_proba (which RF, DT, XGB, LGBM do)
        voting_clf = VotingClassifier(estimators=estimators_for_voting, voting='soft', n_jobs=-1, verbose=1)
        
        print(f"Fitting Voting Classifier with {len(estimators_for_voting)} models...")
        voting_clf.fit(X_train, y_train) # Fit the entire ensemble on original training data

        y_test_pred_ensemble = voting_clf.predict(X_test)
        ensemble_acc = accuracy_score(y_test, y_test_pred_ensemble)
        ensemble_f1 = f1_score(y_test, y_test_pred_ensemble, average='weighted')

        print("\n--- Voting Classifier Test Results ---")
        print(f"Accuracy: {ensemble_acc:.2f}")
        print(f"F1-weighted: {ensemble_f1:.2f}")
        print("Classification Report:")
        print(classification_report(y_test, y_test_pred_ensemble, target_names=class_labels))

        # Confusion Matrix for Ensemble
        fig, ax = plt.subplots(figsize=(8, 6))
        ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_ensemble, display_labels=class_labels,
                                                 cmap='Blues', ax=ax)
        ax.set_title("Confusion Matrix: Voting Classifier (Test Set)")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print("Not enough suitable models for VotingClassifier.")

except Exception as e:
    print(f"Error during Voting Classifier setup or fitting: {e}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, StratifiedKFold
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, f1_score, classification_report, ConfusionMatrixDisplay
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Models
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
import lightgbm as lgb
from sklearn.linear_model import LogisticRegression # For StackingClassifier meta-learner

# Imbalanced-learn
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline # Use imblearn's pipeline for SMOTE integration

# --- 1. โหลดข้อมูล ---
df = pd.read_csv('food_menu.csv') # โหลดไฟล์ CSV ที่เก็บข้อมูลเมนูอาหาร

# Make a copy to avoid SettingWithCopyWarning later if we modify df directly
df_processed = df.copy()

# --- [3] Feature Engineering ---
df_processed['suitable_for_disease'] = LabelEncoder().fit_transform(df_processed['suitable_for_disease'])
# Extract numeric from 'amount'
df_processed['amount_numeric'] = df_processed['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
# Add a small constant (1e-8) to denominators to prevent division by zero, but be aware of its impact on zero values.
df_processed['sugar_to_carb_ratio'] = df_processed['sugar'] / (df_processed['carbohydrate'] + 1e-8)
df_processed['fat_to_calorie_ratio'] = df_processed['fat'] / (df_processed['calories'] + 1e-8)
df_processed['sodium_per_calorie'] = df_processed['sodium'] / (df_processed['calories'] + 1e-8)

# Density & per-calorie features
df_processed['calorie_density'] = df_processed['calories'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['sugar_per_calorie'] = df_processed['sugar'] / (df_processed['calories'] + 1e-8)
df_processed['fat_per_calorie'] = df_processed['fat'] / (df_processed['calories'] + 1e-8)
df_processed['protein_per_calorie'] = df_processed['protein'] / (df_processed['calories'] + 1e-8)
df_processed['carb_per_calorie'] = df_processed['carbohydrate'] / (df_processed['calories'] + 1e-8)
df_processed['protein_to_fat_ratio'] = df_processed['protein'] / (df_processed['fat'] + 1e-8)

# Density by amount
df_processed['protein_density'] = df_processed['protein'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['fat_density'] = df_processed['fat'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['carb_density'] = df_processed['carbohydrate'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['sugar_density'] = df_processed['sugar'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['sodium_density'] = df_processed['sodium'] / (df_processed['amount_numeric'] + 1e-8)

# Extra ratios and densities
df_processed['fiber_per_calorie'] = df_processed['fiber'] / (df_processed['calories'] + 1e-8)
df_processed['fiber_to_carb_ratio'] = df_processed['fiber'] / (df_processed['carbohydrate'] + 1e-8)
df_processed['saturated_fat_per_calorie'] = df_processed['saturated_fat'] / (df_processed['calories'] + 1e-8)
df_processed['saturated_fat_to_total_fat_ratio'] = df_processed['saturated_fat'] / (df_processed['fat'] + 1e-8)
df_processed['fiber_density'] = df_processed['fiber'] / (df_processed['amount_numeric'] + 1e-8)
df_processed['saturated_fat_density'] = df_processed['saturated_fat'] / (df_processed['amount_numeric'] + 1e-8)

# --- **IMPROVEMENT 1: Handle infinite values for sodium_to_sugar_ratio** ---
# Calculate ratio first
df_processed['sodium_to_sugar_ratio'] = df_processed['sodium'] / (df_processed['sugar'] + 1e-8)
# Replace infinite values (which happen when sugar is exactly 0) with 0 or a more suitable value
# Filling with 0 makes sense if no sugar means the ratio is effectively 0 for sodium content relative to sugar.
df_processed['sodium_to_sugar_ratio'] = df_processed['sodium_to_sugar_ratio'].replace([np.inf, -np.inf], 0)
# You could also use: df_processed['sodium_to_sugar_ratio'].fillna(0, inplace=True) if NaNs are introduced elsewhere

# --- **IMPROVEMENT 2: Add more meaningful Interaction Features** ---
df_processed['protein_carb_ratio'] = df_processed['protein'] / (df_processed['carbohydrate'] + 1e-8)
df_processed['nutrient_density_per_amount'] = (df_processed['protein'] + df_processed['fat'] + df_processed['carbohydrate'] + df_processed['fiber']) / (df_processed['amount_numeric'] + 1e-8)
df_processed['sweet_and_salty_index'] = df_processed['sugar'] + df_processed['sodium'] # Simple sum

# --- NEW FEATURE ENGINEERING FOR LOW SUGAR ---
sugar_threshold = df_processed['sugar'].quantile(0.15)
sugar_per_cal_threshold = df_processed['sugar_per_calorie'].quantile(0.15)

df_processed['is_very_low_sugar'] = (df_processed['sugar'] <= sugar_threshold).astype(int)
df_processed['is_very_low_sugar_per_cal'] = (df_processed['sugar_per_calorie'] <= sugar_per_cal_threshold).astype(int)

# --- One-Hot Encode 'category' ---
df_processed = pd.get_dummies(df_processed, columns=['category'], prefix='cat', drop_first=False)

# --- [4] Select Features & Target ---
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie', 'saturated_fat_to_total_fat_ratio',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density',
    'is_very_low_sugar', 'is_very_low_sugar_per_cal', # Add new binary features
    # --- New Interaction Features ---
    'protein_carb_ratio', 'nutrient_density_per_amount', 'sweet_and_salty_index'
]

# Add one-hot encoded category columns
category_ohe_cols = [col for col in df_processed.columns if col.startswith('cat_')]
feature_cols.extend(category_ohe_cols)

X = df_processed[feature_cols]
y = df_processed['food_type']

print("New X dtypes:")
print(X.dtypes)

# Calculate the correlation matrix
correlation_matrix = X.corr()

# Plotting the heatmap
plt.figure(figsize=(20, 18)) # Adjust figure size for better readability with many features
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Features', fontsize=20)
plt.show()
# --- 5. Encode target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print("\nFood Type Mapping:")
for idx, class_label in enumerate(le.classes_):
    print(f"{class_label} --> {idx}")

class_labels = le.classes_

# --- 6. Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)

print(f"\nClass distribution in y_train: {Counter(y_train)}")
print(f"Class distribution in y_test: {Counter(y_test)}")


# --- 7. จัดการ missing values & 8. Scaling & 8.1 SMOTE (สำหรับ train/test evaluation โดยตรง) ---
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# If any class has fewer than k_neighbors + 1 samples, SMOTE will error.
# Check Counter(y_train) for very small classes and adjust k_neighbors if needed.
# For now, keeping at 3.
smote = SMOTE(random_state=42, k_neighbors=3)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

print(f"\nBefore SMOTE: {Counter(y_train)}")
print(f"After SMOTE: {Counter(y_train_balanced)}")
print(f"Training samples (after SMOTE): {X_train_balanced.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")


# --- 9. สร้างโมเดลและกำหนด Hyperparameter Grids สำหรับ Grid Search ---
param_grids = {
    "Random Forest": {
        'classifier': [RandomForestClassifier(random_state=42, class_weight='balanced')],
        'classifier__n_estimators': [150, 300, 500], # Increased max
        'classifier__max_depth': [10, 20, None],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 2],
        'classifier__max_features': ['sqrt', 'log2']
    },
    "KNN": {
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [3, 5, 7, 9, 11, 13], # Reduced range slightly for faster search
        'classifier__weights': ['uniform', 'distance'],
        'classifier__metric': ['euclidean', 'manhattan']
    },
    "Decision Tree": {
        'classifier': [DecisionTreeClassifier(random_state=42, class_weight='balanced')],
        'classifier__max_depth': [8, 15, None],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 2],
        'classifier__max_features': ['sqrt', 'log2']
    },
    "XGBoost": {
        'classifier': [xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss', n_jobs=-1)],
        'classifier__n_estimators': [300, 500, 700], # **IMPROVEMENT 3: Wider range for n_estimators**
        'classifier__learning_rate': [0.01, 0.05, 0.1], # **IMPROVEMENT 3: Finer steps, slightly lower start**
        'classifier__max_depth': [4, 6, 8],
        'classifier__subsample': [0.7, 0.8, 0.9],
        'classifier__colsample_bytree': [0.7, 0.8, 0.9],
        'classifier__gamma': [0, 0.1, 0.2], # Added gamma
        'classifier__reg_alpha': [0, 0.01, 0.1, 0.5] # **IMPROVEMENT 3: Wider range for regularization**
    }
}

best_models = {}
cv_results = {}
kfold_grid = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- Performing Grid Search Cross-Validation ---")
# **IMPROVEMENT 4: Using RandomizedSearchCV for potentially faster tuning with larger grids**
USE_RANDOMIZED_SEARCH = True # Set to False to use GridSearchCV

for name, param_grid in param_grids.items():
    print(f"\nRunning {'Randomized' if USE_RANDOMIZED_SEARCH else 'Grid'} Search for {name}...")

    current_pipeline = ImbPipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42, k_neighbors=3)),
        ('classifier', param_grid['classifier'][0])
    ])

    grid_params = {k: v for k, v in param_grid.items() if k != 'classifier'}

    if USE_RANDOMIZED_SEARCH:
        search_class = RandomizedSearchCV
        search_kwargs = {'n_iter': 70} # Increased n_iter for more comprehensive sampling
        grid_search = search_class(
            estimator=current_pipeline,
            param_distributions=grid_params, # Corrected parameter for RandomizedSearchCV
            cv=kfold_grid,
            scoring='f1_weighted',
            n_jobs=-1,
            verbose=1,
            **search_kwargs
        )
    else:
        search_class = GridSearchCV
        search_kwargs = {}
        grid_search = search_class(
            estimator=current_pipeline,
            param_grid=grid_params, # Corrected parameter for GridSearchCV
            cv=kfold_grid,
            scoring='f1_weighted',
            n_jobs=-1,
            verbose=1,
            **search_kwargs
        )

    grid_search.fit(X_train, y_train)

    best_models[name] = grid_search.best_estimator_
    cv_results[name] = {
        'best_score': grid_search.best_score_,
        'best_params': grid_search.best_params_
    }

    print(f"Finished {'Randomized' if USE_RANDOMIZED_SEARCH else 'Grid'} Search for {name}.")
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best cross-validation F1-weighted: {grid_search.best_score_:.4f}")

# --- 10.1 ประเมินผลบน Train Set (using models from Grid Search) ---
train_accuracies = []
train_f1_scores = []
print("\n--- Evaluating Best Models (from Grid Search) on Train Set ---")
for name, model_pipeline in best_models.items():
    y_train_pred_pipeline = model_pipeline.predict(X_train)
    acc_train = accuracy_score(y_train, y_train_pred_pipeline)
    f1_train = f1_score(y_train, y_train_pred_pipeline, average='weighted', zero_division=0)
    report_train = classification_report(y_train, y_train_pred_pipeline, target_names=class_labels, zero_division=0)

    train_accuracies.append(acc_train)
    train_f1_scores.append(f1_train)

    print(f"\n--- {name} Train Results (Best Model - Pipeline Evaluation) ---")
    print(f"Accuracy: {acc_train:.2f}")
    print(f"F1-weighted: {f1_train:.2f}")
    print("Classification Report:")
    print(report_train)

    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_pipeline, display_labels=class_labels,
                                             cmap='Blues', ax=ax)
    ax.set_title(f"Confusion Matrix: {name} (Train Set - Best Model Pipeline)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- 10.2 ประเมินผลบน Test Set (using models from Grid Search) ---
test_accuracies = []
test_f1_scores = []
model_names = []
print("\n--- Evaluating Best Models (from Grid Search) on Test Set ---")
for name, model_pipeline in best_models.items():
    y_test_pred = model_pipeline.predict(X_test)
    acc = accuracy_score(y_test, y_test_pred)
    f1 = f1_score(y_test, y_test_pred, average='weighted', zero_division=0)
    report = classification_report(y_test, y_test_pred, target_names=class_labels, zero_division=0)

    test_accuracies.append(acc)
    test_f1_scores.append(f1)
    model_names.append(name)

    print(f"\n--- {name} Test Results (Best Model) ---")
    print(f"Accuracy: {acc:.2f}")
    print(f"F1-weighted: {f1:.2f}")
    print("Classification Report:")
    print(report)
    print(f"Best parameters: {cv_results[name]['best_params']}")

    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels,
                                             cmap='Blues', ax=ax)
    ax.set_title(f"Confusion Matrix: {name} (Test Set - Best Model)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- 10.3 แสดงผล Cross Validation ที่ได้จาก Grid Search ---
print("\n--- Cross-Validation Results from Grid Search ---")
for name, gs_result in cv_results.items():
    print(f"{name}: Best CV F1-weighted = {gs_result['best_score']:.3f}, Best Params = {gs_result['best_params']}")

# --- 11.1 กราฟเปรียบเทียบ Accuracy บน Train Set ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=train_accuracies, y=list(best_models.keys()), palette="viridis", orient='h')
for i, v in enumerate(train_accuracies):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Accuracy Comparison on Train Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 11.2 กราฟเปรียบเทียบ F1-weighted บน Train Set ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=train_f1_scores, y=list(best_models.keys()), palette="viridis", orient='h')
for i, v in enumerate(train_f1_scores):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model F1-weighted Comparison on Train Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('F1-weighted Score')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 11.3 กราฟเปรียบเทียบ Accuracy บน Test Set ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=test_accuracies, y=model_names, palette="magma", orient='h')
for i, v in enumerate(test_accuracies):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Accuracy Comparison on Test Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 11.4 กราฟเปรียบเทียบ F1-weighted บน Test Set ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=test_f1_scores, y=model_names, palette="magma", orient='h')
for i, v in enumerate(test_f1_scores):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model F1-weighted Comparison on Test Set (Best Models from Grid Search)', fontsize=16)
plt.xlabel('F1-weighted Score')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()


# --- 11.5 กราฟเปรียบเทียบ Cross Validation F1-weighted จาก Grid Search ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=[res['best_score'] for res in cv_results.values()], y=list(cv_results.keys()), palette="coolwarm", orient='h')
for i, (name, res) in enumerate(cv_results.items()):
    ax.text(res['best_score'] + 0.01, i, f"{res['best_score']:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Best Cross-Validated F1-weighted Score (from Grid Search)', fontsize=16)
plt.xlabel('F1-weighted Score')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.show()

# --- 12. แสดง Feature Importance สำหรับโมเดลที่รองรับ (Random Forest, Decision Tree, XGBoost, LightGBM) ---
feature_names = X.columns.tolist()

for name, model_pipeline in best_models.items():
    classifier = model_pipeline.named_steps['classifier']

    if hasattr(classifier, 'feature_importances_'):
        importances = classifier.feature_importances_
        indices = np.argsort(importances)[::-1]

        top_n = min(20, len(feature_names))
        plt.figure(figsize=(10, max(6, top_n * 0.4)))
        sns.barplot(x=importances[indices[:top_n]], y=np.array(feature_names)[indices[:top_n]], palette='crest')
        plt.title(f"Top {top_n} Feature Importance for {name} (Best Model)")
        plt.xlabel('Importance')
        plt.ylabel('Feature')
        plt.tight_layout()
        plt.show()
    else:
        print(f"\n{name} ไม่รองรับการแสดง Feature Importance (เช่น KNN).")
```
