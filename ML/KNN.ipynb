```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, learning_curve
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay
from collections import Counter # Import Counter for better class distribution visibility
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline # Changed to imblearn.pipeline as it integrates well with SMOTE

# --- 1. Load Data ---
df = pd.read_csv('Food-Menu - food_menu.csv')

# --- 2. Translate food_type labels ---
translate_labels = {
    '': 'Balanced',
    '': 'Low Fat',
    '': 'Low Sugar',
    '': 'Low Sodium'
}
df['food_type'] = df['food_type'].map(translate_labels)

# --- 3. Feature Engineering ---
df['category_encoded'] = LabelEncoder().fit_transform(df['category'])
df['suitable_for_disease'] = LabelEncoder().fit_transform(df['suitable_for_disease'])

df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)
df['sodium_per_calorie'] = df['sodium'] / (df['calories'] + 1e-8)
df['protein_per_calorie'] = df['protein'] / (df['calories'] + 1e-8)

# --- 4. Feature/Target Selection ---
X = df[['calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium',
        'category_encoded', 'suitable_for_disease', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio','sodium_per_calorie','protein_per_calorie']]
y = df['food_type']
print(X.dtypes)

le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_labels = le.classes_
print(f"Food Type Mapping: {list(enumerate(class_labels))}")

# --- 5. Split Train/Evaluation ---
# Calculate combined evaluation set size:
# Original test_size = 0.15
# Original val_size = 0.176 of the remaining (1 - 0.15) = 0.176 * 0.85 = 0.1496
# Combined evaluation_size = 0.15 + 0.1496 = 0.2996 (approx 30%)
eval_set_size = 0.2996 # Adjusted test_size to combine validation and test sets

X_train_raw, X_eval_raw, y_train, y_eval = train_test_split(X, y_encoded, test_size=eval_set_size, stratify=y_encoded, random_state=42)

print(f"\nTraining set size: {len(X_train_raw)}")
print(f"Evaluation set size: {len(X_eval_raw)}")

# --- 6. Preprocessing (Impute and Scale) ---
imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()

# Fit imputer and scaler only on the training data, then transform both train and eval sets
X_train_processed = scaler.fit_transform(imputer.fit_transform(X_train_raw))
X_eval_processed = scaler.transform(imputer.transform(X_eval_raw))

# --- 7. SMOTE (Applied only to the training set) ---
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_processed, y_train)

print(f"\nClass distribution in y_train (before SMOTE): {Counter(y_train)}")
print(f"Class distribution in y_train (after SMOTE): {Counter(y_train_balanced)}")
print(f"Class distribution in y_eval: {Counter(y_eval)}")


# --- 8. Models ---
models = {
    "Random Forest": RandomForestClassifier(
        n_estimators=50,
        max_depth=5,
        min_samples_split=5,
        min_samples_leaf=5,
        random_state=42
    ),
    "KNN": KNeighborsClassifier(metric='manhattan', n_neighbors=5, weights='distance'),
    "Decision Tree": DecisionTreeClassifier(max_depth=10, min_samples_leaf=4, min_samples_split=4, max_features='log2', class_weight='balanced', random_state=42)
}

trained_models = {}
train_accuracies, eval_accuracies, model_names = [], [], [] # Renamed test_accuracies to eval_accuracies, removed val_accuracies

# --- 9. Train & Evaluate ---
print("\n--- Training Models ---")
for name, model in models.items():
    model.fit(X_train_balanced, y_train_balanced)
    trained_models[name] = model

    # Train eval
    y_pred_train = model.predict(X_train_balanced)
    acc_train = accuracy_score(y_train_balanced, y_pred_train)
    train_accuracies.append(acc_train)
    print(f"\n{name} Train Accuracy: {acc_train:.2f}")
    print(classification_report(y_train_balanced, y_pred_train, target_names=class_labels))
    ConfusionMatrixDisplay.from_predictions(y_train_balanced, y_pred_train, display_labels=class_labels, cmap='Blues')
    plt.title(f"{name} - Confusion Matrix (Train)")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Evaluation eval (now combines original Validation and Test sets)
    y_pred_eval = model.predict(X_eval_processed)
    acc_eval = accuracy_score(y_eval, y_pred_eval)
    eval_accuracies.append(acc_eval)
    model_names.append(name) # Capture model names for plotting
    print(f"\n{name} Test Accuracy: {acc_eval:.2f}")
    print(classification_report(y_eval, y_pred_eval, target_names=class_labels))
    ConfusionMatrixDisplay.from_predictions(y_eval, y_pred_eval, display_labels=class_labels, cmap='Blues')
    plt.title(f"{name} - Confusion Matrix (Test)")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# --- 10. Accuracy Comparison Plots ---
def plot_bar_with_values(x, y, title, color):
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=x, y=y, palette=color, orient='h')
    for i, v in enumerate(x):
        ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
    plt.title(title)
    plt.xlabel('Accuracy')
    plt.xlim(0, 1)
    plt.tight_layout()
    plt.show()

plot_bar_with_values(train_accuracies, list(models.keys()), 'Train Accuracy', 'viridis')
plot_bar_with_values(eval_accuracies, model_names, 'Test Accuracy', 'magma') # Plot for combined evaluation set

# --- 11. Cross Validation ---
cv_scores = {}
cv_stddev = {}
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- Cross-Validation Results ---")
for name, base_model in models.items():
    # The pipeline for cross-validation will handle preprocessing and SMOTE internally for each fold.
    pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)),
        ('classifier', base_model)
    ])
    scores = cross_val_score(pipe, X, y_encoded, cv=kfold, scoring='accuracy')
    cv_scores[name] = scores.mean()
    cv_stddev[name] = scores.std()
    print(f"{name}: CV Mean Accuracy = {scores.mean():.3f}, Std = {scores.std():.3f}")

plot_bar_with_values(list(cv_scores.values()), list(cv_scores.keys()), 'Cross-Validated Accuracy', 'coolwarm')

# --- 12. Feature Importance ---
feature_names = X.columns.tolist() # Use X.columns for original feature names
for name, model in trained_models.items():
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(10, 6))
        sns.barplot(x=importances[indices], y=np.array(feature_names)[indices], palette='crest')
        plt.title(f"Feature Importance - {name}")
        plt.tight_layout()
        plt.show()
    else:
        print(f"{name} does not support feature importance.")

# --- 13. Learning Curve for All Models ---
def plot_learning_curve(model, model_name, X, y, cv, scoring='accuracy'):
    # The pipeline for learning curve will also handle preprocessing and SMOTE internally.
    pipe = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])

    train_sizes, train_scores, val_scores = learning_curve(
        estimator=pipe,
        X=X, # Use the full dataset here, as learning_curve performs internal CV splits
        y=y,
        train_sizes=np.linspace(0.1, 1.0, 5),
        cv=cv,
        scoring=scoring,
        n_jobs=-1,
        shuffle=True,
        random_state=42
    )

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_mean, 'o-', color='green', label='Training Accuracy')
    plt.plot(train_sizes, val_mean, 'o-', color='blue', label='Cross-validation Accuracy') # Label changed
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='green')
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='blue')
    for i in range(len(train_sizes)):
        plt.text(train_sizes[i], train_mean[i], f"{train_mean[i]:.2f}", color='green', ha='right')
        plt.text(train_sizes[i], val_mean[i], f"{val_mean[i]:.2f}", color='blue', ha='left')
    plt.title(f'Learning Curve: {model_name}')
    plt.xlabel('Number of Training Examples')
    plt.ylabel(scoring.capitalize())
    plt.legend(loc='best')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Call learning curve for all models
for name, model in models.items():
    plot_learning_curve(model, name, X, y_encoded, cv=kfold) # Using the same kfold instance
```
