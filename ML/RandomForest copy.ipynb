```python
# --- [1] Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, learning_curve
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

# --- [2] Load Dataset ---
df = pd.read_csv('food_menu.csv')

# --- [3] Feature Engineering ---
# Encode category
df['category_encoded'] = LabelEncoder().fit_transform(df['category'])

# Extract numeric from 'amount'
df['amount_numeric'] = df['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)
df['sodium_per_calorie'] = df['sodium'] / (df['calories'] + 1e-8)

# Density & per-calorie features
df['calorie_density'] = df['calories'] / (df['amount_numeric'] + 1e-8)
df['sugar_per_calorie'] = df['sugar'] / (df['calories'] + 1e-8)
df['fat_per_calorie'] = df['fat'] / (df['calories'] + 1e-8)
df['protein_per_calorie'] = df['protein'] / (df['calories'] + 1e-8)
df['carb_per_calorie'] = df['carbohydrate'] / (df['calories'] + 1e-8)
df['protein_to_fat_ratio'] = df['protein'] / (df['fat'] + 1e-8)

# Density by amount
df['protein_density'] = df['protein'] / (df['amount_numeric'] + 1e-8)
df['fat_density'] = df['fat'] / (df['amount_numeric'] + 1e-8)
df['carb_density'] = df['carbohydrate'] / (df['amount_numeric'] + 1e-8)
df['sugar_density'] = df['sugar'] / (df['amount_numeric'] + 1e-8)
df['sodium_density'] = df['sodium'] / (df['amount_numeric'] + 1e-8)

# Extra ratios and densities
df['fiber_per_calorie'] = df['fiber'] / (df['calories'] + 1e-8)
df['fiber_to_carb_ratio'] = df['fiber'] / (df['carbohydrate'] + 1e-8)
df['sodium_to_sugar_ratio'] = df['sodium'] / (df['sugar'] + 1e-8)
df['saturated_fat_per_calorie'] = df['saturated_fat'] / (df['calories'] + 1e-8)
df['saturated_fat_to_total_fat_ratio'] = df['saturated_fat'] / (df['fat'] + 1e-8)
df['fiber_density'] = df['fiber'] / (df['amount_numeric'] + 1e-8)
df['saturated_fat_density'] = df['saturated_fat'] / (df['amount_numeric'] + 1e-8)

# --- Feature Engineering เพิ่มเติมสำหรับคลาส 'น้ำตาลต่ำ' และ 'ไขมันต่ำ' ---

# 1. คุณสมบัติแบบ Interaction (การรวมคุณสมบัติ)
# การคูณหรือหารคุณสมบัติที่เกี่ยวข้อง อาจช่วยขยายความแตกต่าง
df['sugar_x_fat'] = df['sugar'] * df['fat']
df['sugar_plus_carb'] = df['sugar'] + df['carbohydrate'] # สัดส่วนรวมของพลังงานจากน้ำตาลและคาร์โบไฮเดรต
df['fat_plus_saturated_fat'] = df['fat'] + df['saturated_fat'] # ไขมันรวมที่สำคัญ

# 2. คุณสมบัติแบบ Polynomial (การยกกำลัง - อาจช่วยให้เส้นแบ่งคลาสไม่เป็นเชิงเส้น)
# df['sugar_sq'] = df['sugar']**2 # ถ้าต้องการเพิ่มความโดดเด่นของค่าน้ำตาลที่สูง
# df['fat_sq'] = df['fat']**2 # ถ้าต้องการเพิ่มความโดดเด่นของค่าไขมันที่สูง
# df['sugar_sqrt'] = np.sqrt(df['sugar']) # ถ้าต้องการลดผลกระทบของค่าสูงๆ และดึงความแตกต่างของค่าต่ำๆ (สำหรับน้ำตาลต่ำ)
# df['fat_sqrt'] = np.sqrt(df['fat']) # คล้ายกันสำหรับไขมันต่ำ
# *ข้อควรระวัง: Polynomial Features อาจทำให้เกิด Multicollinearity และ Overfitting ได้ง่าย ต้องระวัง*

# 3. คุณสมบัติเชิง Threshold (อ้างอิงจากเกณฑ์ที่คาดว่าเกี่ยวข้องกับ 'ต่ำ')
# เนื่องจากเราไม่มีเกณฑ์ที่ชัดเจนว่า "น้ำตาลต่ำ" หรือ "ไขมันต่ำ" คือเท่าไหร่
# เราสามารถใช้ค่า Median หรือ Quantile ของ 'sugar'/'fat' จากข้อมูลทั้งหมด หรือจากคลาสที่มีปัญหา
# มาสร้างเป็นเกณฑ์เบื้องต้น

# คำนวณ Median หรือ Quantile จากข้อมูลทั้งหมด
median_sugar = df['sugar'].median()
median_fat = df['fat'].median()
q25_sugar = df['sugar'].quantile(0.25) # 25th percentile
q25_fat = df['fat'].quantile(0.25) # 25th percentile

# สร้างคุณสมบัติ Binary (0 หรือ 1) โดยใช้เกณฑ์
# is_low_sugar: 1 ถ้า 'sugar' ต่ำกว่าค่า Median/Quantile
df['is_low_sugar_median'] = (df['sugar'] < median_sugar).astype(int)
df['is_low_sugar_q25'] = (df['sugar'] < q25_sugar).astype(int)

# is_low_fat: 1 ถ้า 'fat' ต่ำกว่าค่า Median/Quantile
df['is_low_fat_median'] = (df['fat'] < median_fat).astype(int)
df['is_low_fat_q25'] = (df['fat'] < q25_fat).astype(int)

# อาจลองสร้างคุณสมบัติรวมของสารอาหารอื่นๆ ที่ไม่เกี่ยวข้อง เพื่อแยกคลาส "สมดุล" ออกไป
df['total_macro_nutrients'] = df['protein'] + df['carbohydrate'] + df['fat']

# คุณสมบัติที่เน้นความแตกต่างจากค่าเฉลี่ย/Median ของคลาสอื่น
# จำเป็นต้องมี `food_type` เพื่อคำนวณค่าเฉลี่ยของคลาสอื่นๆ
# เนื่องจากเราต้องการแยก 'น้ำตาลต่ำ' และ 'ไขมันต่ำ'
# ให้ลองคำนวณค่าเฉลี่ยของน้ำตาลและไขมัน **จากคลาสที่ไม่ใช่น้ำตาลต่ำ/ไขมันต่ำ**
# เพื่อใช้เป็นฐานในการเปรียบเทียบ

# ขั้นแรก: กำหนด Class Labels (ถ้ายังไม่มี)
# ต้องแน่ใจว่าได้รันส่วน LabelEncoder(y) มาแล้ว เพื่อให้รู้ว่า 'น้ำตาลต่ำ' และ 'ไขมันต่ำ' คือ index อะไร
# สมมติว่า 'น้ำตาลต่ำ' คือ index 0, 'ไขมันต่ำ' คือ index 3 (ตรวจสอบจาก print("Class Mapping:") ก่อนหน้านี้)
# ต้องใช้ y_encoded ในการคำนวณ
# เนื่องจากโค้ดนี้รันก่อน y_encoded จะถูกสร้าง, ต้องทำการ map y กลับไปก่อน หรือคำนวณหลังจาก y_encoded ถูกสร้าง
# สำหรับตอนนี้ ขอสมมติว่า 'น้ำตาลต่ำ' และ 'ไขมันต่ำ' เป็นค่าในคอลัมน์ 'food_type' โดยตรง
# ถ้าคุณมี `food_type` เป็น string ใน df ให้ใช้แบบนี้
# avg_sugar_not_low_sugar = df[df['food_type'] != 'น้ำตาลต่ำ']['sugar'].mean()
# avg_fat_not_low_fat = df[df['food_type'] != 'ไขมันต่ำ']['fat'].mean()

# เพื่อให้โค้ด Feature Engineering อยู่ในบล็อกเดียวกันกับโค้ดที่คุณให้มา
# เราจะคำนวณค่าเฉลี่ยเหล่านี้หลังจากที่ `food_type` (original string) มีอยู่
# โดยอาจจะต้องรันโค้ดส่วน [2] และ [3] ใหม่ทั้งหมด
# หรือให้ข้อมูลเหล่านี้มาเพิ่ม
# แต่ถ้าต้องการสร้าง FE ที่ซับซ้อนขึ้น ควรทำในขั้นตอนแยกต่างหาก หรือหลัง y_encoded ถูกสร้าง
# **สำหรับตอนนี้ เราจะข้ามส่วนนี้ไปก่อน หากข้อมูล Food_type ที่แท้จริงไม่ได้อยู่ใน df นี้โดยตรง**
# (เนื่องจากในส่วน [4] ถึงจะมีการกำหนด y = df['food_type'] และ Encode)
# ดังนั้น เราจะเน้นไปที่คุณสมบัติที่สร้างได้จากค่าคุณสมบัติเท่านั้นก่อน

# --- [4] Select Features & Target ---
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie', 'saturated_fat_to_total_fat_ratio',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density', 'category_encoded',
    # เพิ่มคุณสมบัติใหม่ที่สร้างขึ้น
    'sugar_x_fat',
    'sugar_plus_carb',
    'fat_plus_saturated_fat',
    'is_low_sugar_median',
    'is_low_sugar_q25',
    'is_low_fat_median',
    'is_low_fat_q25',
    'total_macro_nutrients'
    # ถ้ามีการเพิ่ม Polynomial Features ต้องเพิ่มตรงนี้ด้วย เช่น 'sugar_sq', 'fat_sq'
]
X = df[feature_cols]
y = df['food_type']

print("--- New Features Added for targeted separation: ---")
print("sugar_x_fat: Product of sugar and fat")
print("sugar_plus_carb: Sum of sugar and carbohydrate")
print("fat_plus_saturated_fat: Sum of fat and saturated fat")
print("is_low_sugar_median: Binary flag if sugar is below dataset median")
print("is_low_sugar_q25: Binary flag if sugar is below dataset 25th percentile")
print("is_low_fat_median: Binary flag if fat is below dataset median")
print("is_low_fat_q25: Binary flag if fat is below dataset 25th percentile")
print("total_macro_nutrients: Sum of protein, carbohydrate, and fat")
print(f"\nTotal features now: {len(feature_cols)}")

# --- [5] Encode Target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_labels = le.classes_

print("Class Mapping:")
for idx, name in enumerate(class_labels):
    print(f"{name} --> {idx}")

# --- [6] Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

# --- [7] Build Pipeline ---
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(
        n_estimators=50,
        max_depth=10,
        min_samples_split=2,
        min_samples_leaf=2,
        random_state=42,
        class_weight='balanced'
    ))
])

# --- [8] Train Pipeline ---
pipeline.fit(X_train, y_train)

# --- [9] Evaluate Train Set ---
y_train_pred = pipeline.predict(X_train)
print("\n📊 Train Set Evaluation:")
print(f"Accuracy: {accuracy_score(y_train, y_train_pred):.2f}")
print(classification_report(y_train, y_train_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Train Set")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- [10] Evaluate Test Set ---
y_test_pred = pipeline.predict(X_test)
print("\n📊 Test Set Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.2f}")
print(classification_report(y_test, y_test_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Test Set")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# --- [11] Feature Importance ---
# Bypass Pipeline (SMOTE affects train only)
X_train_imp = SimpleImputer(strategy='mean').fit_transform(X_train)
X_train_scaled = StandardScaler().fit_transform(X_train_imp)
X_bal, y_bal = SMOTE(random_state=42).fit_resample(X_train_scaled, y_train)

rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X_bal, y_bal)

importances = rf.feature_importances_
sorted_idx = np.argsort(importances)[::-1]
sorted_features = np.array(feature_cols)[sorted_idx]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[sorted_idx], y=sorted_features, palette='crest')
plt.title("Feature Importance")
plt.tight_layout()
plt.show()

# --- [12] Learning Curve ---
train_sizes, train_scores, test_scores = learning_curve(
    rf, X_bal, y_bal, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1
)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve")
plt.legend()
plt.tight_layout()
plt.show()

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for label in np.unique(y):
    plt.scatter(X_pca[y == label, 0], X_pca[y == label, 1], label=label, alpha=0.6)
plt.title("PCA: Class Separation")
plt.legend()
plt.show()


from sklearn.tree import plot_tree

# เลือกต้นไม้ต้นที่ 0 จาก Random Forest ที่ train ไปแล้ว
tree_to_plot = rf.estimators_[0]  # เลือกต้นไม้ต้นที่ 1 (index 0)

# สร้างแผนผัง
plt.figure(figsize=(24, 12))
plot_tree(tree_to_plot,
          feature_names=feature_cols,
          class_names=class_labels,
          filled=True,
          rounded=True,
          fontsize=8,
          max_depth=3  # ปรับเพื่อให้แสดงแค่ระดับบนๆ
         )

plt.title("🌳 Random Forest: Tree Visualization (Tree #0, max_depth=3)")
plt.tight_layout()
plt.show()


```

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
import matplotlib.pyplot as plt

# 1. เตรียม classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# 2. Stratified CV เพื่อให้กระจายแต่ละ class เท่ากันในแต่ละ fold
cv = StratifiedKFold(n_splits=5)

# 3. สร้าง RFECV object
selector = RFECV(estimator=rf, step=1, cv=cv, scoring='accuracy', n_jobs=-1)

# 4. Fit ลงไปกับข้อมูล
selector.fit(X, y)

# 5. แสดงผลลัพธ์
print(f"Optimal number of features: {selector.n_features_}")
print("Selected features:")
selected_features = X.columns[selector.support_]
print(selected_features)

# 6. Plot graph
plt.figure(figsize=(10, 6))
plt.plot(
    range(1, len(selector.cv_results_['mean_test_score']) + 1),
    selector.cv_results_['mean_test_score'],
    marker='o'
)
plt.xlabel("Number of Features Selected")
plt.ylabel("Cross-Validation Accuracy")
plt.title("Feature Selection with RFECV")
plt.grid(True)
plt.show()

rfecv_scores = selector.cv_results_['mean_test_score']
for i, score in enumerate(rfecv_scores, 1):
    print(f"{i} features: {score:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# --- [2] Load Dataset --- (โหลดข้อมูลอีกครั้งเพื่อให้โค้ดนี้รันได้แยก)
df = pd.read_csv('food_menu.csv')

# --- [3] Feature Engineering --- (ทำ Feature Engineering อีกครั้ง)
# Encode category
df['category_encoded'] = LabelEncoder().fit_transform(df['category'])

# Extract numeric from 'amount'
df['amount_numeric'] = df['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)
df['sodium_per_calorie'] = df['sodium'] / (df['calories'] + 1e-8)

# Density & per-calorie features
df['calorie_density'] = df['calories'] / (df['amount_numeric'] + 1e-8)
df['sugar_per_calorie'] = df['sugar'] / (df['calories'] + 1e-8)
df['fat_per_calorie'] = df['fat'] / (df['calories'] + 1e-8)
df['protein_per_calorie'] = df['protein'] / (df['calories'] + 1e-8)
df['carb_per_calorie'] = df['carbohydrate'] / (df['calories'] + 1e-8)
df['protein_to_fat_ratio'] = df['protein'] / (df['fat'] + 1e-8)

# Density by amount
df['protein_density'] = df['protein'] / (df['amount_numeric'] + 1e-8)
df['fat_density'] = df['fat'] / (df['amount_numeric'] + 1e-8)
df['carb_density'] = df['carbohydrate'] / (df['amount_numeric'] + 1e-8)
df['sugar_density'] = df['sugar'] / (df['amount_numeric'] + 1e-8)
df['sodium_density'] = df['sodium'] / (df['amount_numeric'] + 1e-8)

# Extra ratios and densities
df['fiber_per_calorie'] = df['fiber'] / (df['calories'] + 1e-8)
df['fiber_to_carb_ratio'] = df['fiber'] / (df['carbohydrate'] + 1e-8)
df['sodium_to_sugar_ratio'] = df['sodium'] / (df['sugar'] + 1e-8)
df['saturated_fat_per_calorie'] = df['saturated_fat'] / (df['calories'] + 1e-8)
df['saturated_fat_to_total_fat_ratio'] = df['saturated_fat'] / (df['fat'] + 1e-8)
df['fiber_density'] = df['fiber'] / (df['amount_numeric'] + 1e-8)
df['saturated_fat_density'] = df['saturated_fat'] / (df['amount_numeric'] + 1e-8)

# --- [4] Select Features & Target ---
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie', 'saturated_fat_to_total_fat_ratio',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density', 'category_encoded'
]
X = df[feature_cols]
y = df['food_type']

# --- [5] Encode Target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_labels = le.classes_ # ['น้ำตาลต่ำ', 'สมดุล', 'โซเดียมต่ำ', 'ไขมันต่ำ'] (ตรวจสอบลำดับจาก output เก่า)

print(f"Class labels mapping: {list(enumerate(class_labels))}")

# --- EDA สำหรับคลาสที่มีปัญหา ---

# กำหนดคลาสที่มีปัญหาและคุณสมบัติที่เกี่ยวข้อง
problem_classes = ['น้ำตาลต่ำ', 'ไขมันต่ำ']
# คุณสมบัติที่น่าจะเกี่ยวข้องโดยตรง
sugar_related_features = ['sugar', 'sugar_density', 'sugar_per_calorie', 'sugar_to_carb_ratio', 'sodium_to_sugar_ratio']
fat_related_features = ['fat', 'fat_density', 'fat_per_calorie', 'fat_to_calorie_ratio', 'saturated_fat', 'saturated_fat_density', 'saturated_fat_to_total_fat_ratio']
# คุณสมบัติสำคัญจาก Feature Importance (เลือกบางส่วน)
top_features_from_importance = ['saturated_fat_density', 'calorie_density', 'sodium', 'carbohydrate', 'sodium_density', 'protein']

all_relevant_features = list(set(sugar_related_features + fat_related_features + top_features_from_importance))


print("\n--- 1. Box Plot / Violin Plot for Key Features vs. Food Type ---")
# พล็อต Box Plot/Violin Plot สำหรับคุณสมบัติที่เกี่ยวข้องโดยตรงและคุณสมบัติสำคัญ
for feature in all_relevant_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='food_type', y=feature, data=df, palette='viridis')
    plt.title(f'Distribution of {feature} by Food Type')
    plt.xlabel('Food Type')
    plt.ylabel(feature)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# อาจลอง Violin plot ด้วยเพื่อดูการกระจายตัวของข้อมูล
# for feature in all_relevant_features:
#     plt.figure(figsize=(10, 6))
#     sns.violinplot(x='food_type', y=feature, data=df, palette='viridis')
#     plt.title(f'Distribution of {feature} by Food Type (Violin Plot)')
#     plt.xlabel('Food Type')
#     plt.ylabel(feature)
#     plt.xticks(rotation=45, ha='right')
#     plt.tight_layout()
#     plt.show()


print("\n--- 2. Correlation Matrix ---")
# รวมคุณสมบัติทั้งหมดและ target เข้าใน DataFrame เดียวกันสำหรับการคำนวณ Correlation
df_corr = df[feature_cols].copy()
df_corr['food_type_encoded'] = y_encoded # ใช้ y_encoded เพื่อดูความสัมพันธ์

plt.figure(figsize=(16, 14))
sns.heatmap(df_corr.corr(), annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5) # annot=True อาจทำให้เยอะเกินไป
plt.title('Correlation Matrix of Features and Encoded Food Type')
plt.tight_layout()
plt.show()

# แสดง Correlation ของคุณสมบัติกับ food_type_encoded (target) โดยเฉพาะ
print("\nCorrelation of Features with Encoded Food Type:")
print(df_corr.corr()['food_type_encoded'].sort_values(ascending=False))


print("\n--- 3. Pair Plot for Selected Key Features ---")
# เลือกคุณสมบัติจำนวนน้อยที่สำคัญจริงๆ เพื่อทำ Pair Plot (เพราะถ้าเยอะจะแสดงผลไม่ได้)
# เลือกจาก Top Features และที่เกี่ยวข้องกับ Sugar/Fat
selected_pair_features = ['sugar', 'fat', 'fiber_per_calorie', 'calorie_density', 'protein', 'food_type']
# ตรวจสอบว่าคอลัมน์ทั้งหมดมีอยู่ใน df_corr ก่อนทำ pairplot
existing_pair_features = [col for col in selected_pair_features if col in df.columns]
if 'food_type' not in existing_pair_features:
    existing_pair_features.append('food_type') # ตรวจสอบให้แน่ใจว่ามี food_type

plt.figure(figsize=(12, 12))
sns.pairplot(df[existing_pair_features], hue='food_type', palette='viridis', diag_kind='kde', plot_kws={'alpha': 0.7})
plt.suptitle('Pair Plot of Key Features by Food Type', y=1.02) # y คือตำแหน่งของ title
plt.tight_layout(rect=[0, 0, 1, 0.98]) # ปรับ layout เพื่อให้ title ไม่ทับ
plt.show()

print("\n--- EDA Completed ---")
print("โปรดดูภาพกราฟที่สร้างขึ้นเพื่อวิเคราะห์ว่าคลาส 'น้ำตาลต่ำ' และ 'ไขมันต่ำ' แตกต่างจากคลาสอื่นอย่างไรในแต่ละคุณสมบัติ")
```

```python
# --- [1] Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, learning_curve, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay, make_scorer, f1_score
from sklearn.decomposition import PCA
from sklearn.tree import plot_tree

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.utils import class_weight

# --- [2] Load Dataset ---
df = pd.read_csv('food_menu.csv')

# --- [3] Feature Engineering (ใช้ชุดล่าสุดที่ลบ 'saturated_fat_to_total_fat_ratio' แล้ว) ---
# Encode category
df['category_encoded'] = LabelEncoder().fit_transform(df['category'])

# Extract numeric from 'amount'
df['amount_numeric'] = df['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)
df['sodium_per_calorie'] = df['sodium'] / (df['calories'] + 1e-8)

# Density & per-calorie features
df['calorie_density'] = df['calories'] / (df['amount_numeric'] + 1e-8)
df['sugar_per_calorie'] = df['sugar'] / (df['calories'] + 1e-8)
df['fat_per_calorie'] = df['fat'] / (df['calories'] + 1e-8)
df['protein_per_calorie'] = df['protein'] / (df['calories'] + 1e-8)
df['carb_per_calorie'] = df['carbohydrate'] / (df['calories'] + 1e-8)
df['protein_to_fat_ratio'] = df['protein'] / (df['fat'] + 1e-8)

# Density by amount
df['protein_density'] = df['protein'] / (df['amount_numeric'] + 1e-8)
df['fat_density'] = df['fat'] / (df['amount_numeric'] + 1e-8)
df['carb_density'] = df['carbohydrate'] / (df['amount_numeric'] + 1e-8)
df['sugar_density'] = df['sugar'] / (df['amount_numeric'] + 1e-8)
df['sodium_density'] = df['sodium'] / (df['amount_numeric'] + 1e-8)

# Extra ratios and densities
df['fiber_per_calorie'] = df['fiber'] / (df['calories'] + 1e-8)
df['fiber_to_carb_ratio'] = df['fiber'] / (df['carbohydrate'] + 1e-8)
df['sodium_to_sugar_ratio'] = df['sodium'] / (df['sugar'] + 1e-8)
df['saturated_fat_per_calorie'] = df['saturated_fat'] / (df['calories'] + 1e-8)
df['fiber_density'] = df['fiber'] / (df['amount_numeric'] + 1e-8)
df['saturated_fat_density'] = df['saturated_fat'] / (df['amount_numeric'] + 1e-8)

# Interaction Features
df['sugar_x_fat'] = df['sugar'] * df['fat']
df['sugar_plus_carb'] = df['sugar'] + df['carbohydrate']
df['fat_plus_saturated_fat'] = df['fat'] + df['saturated_fat']
df['total_macro_nutrients'] = df['protein'] + df['carbohydrate'] + df['fat']

# Binary Features (คงไว้ตามเดิม)
median_sugar = df['sugar'].median()
median_fat = df['fat'].median()
q25_sugar = df['sugar'].quantile(0.25)
q25_fat = df['fat'].quantile(0.25)

df['is_low_sugar_median'] = (df['sugar'] < median_sugar).astype(int)
df['is_low_sugar_q25'] = (df['sugar'] < q25_sugar).astype(int)
df['is_low_fat_median'] = (df['fat'] < median_fat).astype(int)
df['is_low_fat_q25'] = (df['fat'] < q25_fat).astype(int)


# RDI-based Features
RDI_CALORIES = 2000
RDI_SUGAR = 25
RDI_FAT = 65
RDI_SODIUM = 2000
RDI_PROTEIN = 50
RDI_CARB = 275
RDI_FIBER = 28
RDI_SAT_FAT = 20

df['sugar_rdi_ratio'] = df['sugar'] / RDI_SUGAR
df['fat_rdi_ratio'] = df['fat'] / RDI_FAT
df['sodium_rdi_ratio'] = df['sodium'] / RDI_SODIUM
df['protein_rdi_ratio'] = df['protein'] / RDI_PROTEIN
df['carb_rdi_ratio'] = df['carbohydrate'] / RDI_CARB
df['fiber_rdi_ratio'] = df['fiber'] / RDI_FIBER
df['sat_fat_rdi_ratio'] = df['saturated_fat'] / RDI_SAT_FAT
df['calories_rdi_ratio'] = df['calories'] / RDI_CALORIES

df['is_high_sugar_rdi'] = (df['sugar_rdi_ratio'] > (1/3)).astype(int)
df['is_low_sugar_rdi'] = (df['sugar_rdi_ratio'] < (1/25)).astype(int)

df['is_high_fat_rdi'] = (df['fat_rdi_ratio'] > (1/3)).astype(int)
df['is_low_fat_rdi'] = (df['fat_rdi_ratio'] < (1/25)).astype(int)

df['is_high_sodium_rdi'] = (df['sodium_rdi_ratio'] > (1/3)).astype(int)
df['is_low_sodium_rdi'] = (df['sodium_rdi_ratio'] < (1/25)).astype(int)

df['is_high_sat_fat_rdi'] = (df['sat_fat_rdi_ratio'] > (1/3)).astype(int)

# Specific Threshold Features (คงไว้)
df['is_very_low_sugar'] = (df['sugar'] < 5).astype(int)
df['is_zero_sugar'] = (df['sugar'] == 0).astype(int)

df['is_very_low_fat'] = (df['fat'] < 3).astype(int)
df['is_fat_free'] = (df['fat'] == 0).astype(int)

df['is_fda_low_sodium'] = (df['sodium'] <= 140).astype(int)


# --- [4] Select Features & Target ---
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density', 'category_encoded',
    'sugar_x_fat',
    'sugar_plus_carb',
    'fat_plus_saturated_fat',
    'is_low_sugar_median',
    'is_low_sugar_q25',
    'is_low_fat_median',
    'is_low_fat_q25',
    'total_macro_nutrients',
    # RDI-based Features
    'sugar_rdi_ratio', 'fat_rdi_ratio', 'sodium_rdi_ratio', 'protein_rdi_ratio',
    'carb_rdi_ratio', 'fiber_rdi_ratio', 'sat_fat_rdi_ratio', 'calories_rdi_ratio',
    'is_high_sugar_rdi', 'is_low_sugar_rdi', 'is_high_fat_rdi', 'is_low_fat_rdi',
    'is_high_sodium_rdi', 'is_low_sodium_rdi', 'is_high_sat_fat_rdi',
    # Specific Threshold Features
    'is_very_low_sugar', 'is_zero_sugar', 'is_very_low_fat', 'is_fat_free', 'is_fda_low_sodium'
]
X = df[feature_cols]
y = df['food_type']

# --- [5] Encode Target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_labels = le.classes_

print("Class Mapping:")
for idx, name in enumerate(class_labels):
    print(f"{name} --> {idx}")

# --- [6] Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

# --- [7] Build Pipeline (สำหรับ GridSearchCV) ---
classes = np.unique(y_train)
weights = class_weight.compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights_dict = dict(zip(classes, weights))

print("\nCalculated Class Weights (initial balanced):")
for cls_idx, weight in class_weights_dict.items():
    print(f"Class {le.inverse_transform([cls_idx])[0]} (Index {cls_idx}): {weight:.2f}")

# **ปรับ Class Weights ด้วยมือสำหรับ 'ไขมันต่ำ' และ 'น้ำตาลต่ำ' (กลับไปใช้ 1.2x)**
# ดึง index ของคลาส 'ไขมันต่ำ' และ 'น้ำตาลต่ำ'
fat_low_idx = le.transform(['ไขมันต่ำ'])[0]
sugar_low_idx = le.transform(['น้ำตาลต่ำ'])[0]

# กลับไปใช้ 1.2 เท่าของน้ำหนักที่คำนวณได้จาก 'balanced'
class_weights_dict[fat_low_idx] = class_weights_dict[fat_low_idx] * 1.2
class_weights_dict[sugar_low_idx] = class_weights_dict[sugar_low_idx] * 1.2

print("\nAdjusted Class Weights (reverted to 1.2x boost):")
for cls_idx, weight in class_weights_dict.items():
    print(f"Class {le.inverse_transform([cls_idx])[0]} (Index {cls_idx}): {weight:.2f}")

base_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42, class_weight=class_weights_dict)) # ใช้ class_weight ที่ปรับแล้ว
])

# --- [8] ใช้ Best Parameters ที่ได้จากการรันก่อนหน้านี้ ---
print("\n--- Model Training with Adjusted Class Weights (Reverted to 1.2x Boost) ---")

# Best Parameters จากการรันครั้งล่าสุด (ที่ได้ Accuracy 0.80)
best_params_for_classifier = {
    'criterion': 'gini',
    'max_depth': 15,
    'max_features': 0.7,
    'min_samples_leaf': 3,
    'min_samples_split': 10,
    'n_estimators': 350
}

final_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42, class_weight=class_weights_dict, **best_params_for_classifier))
])

final_pipeline.fit(X_train, y_train)

print("\n--- Model Training with Adjusted Class Weights (Reverted to 1.2x Boost) Completed ---")
print(f"Parameters used for classifier: {final_pipeline.named_steps['classifier'].get_params()}")


# --- [9] Evaluate Train Set (with Best Model) ---
y_train_pred = final_pipeline.predict(X_train)
print("\n📊 Train Set Evaluation (Best Model with Adjusted Class Weights - 1.2x Boost):")
print(f"Accuracy: {accuracy_score(y_train, y_train_pred):.2f}")
print(classification_report(y_train, y_train_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Train Set (Best Model with Adjusted Class Weights - 1.2x Boost)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- [10] Evaluate Test Set (with Best Model) ---
y_test_pred = final_pipeline.predict(X_test)
print("\n📊 Test Set Evaluation (Best Model with Adjusted Class Weights - 1.2x Boost):")
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.2f}")
print(classification_report(y_test, y_test_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Test Set (Best Model with Adjusted Class Weights - 1.2x Boost)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# --- [11] Feature Importance (จาก Best Model) ---
best_rf_classifier = final_pipeline.named_steps['classifier']

if hasattr(best_rf_classifier, 'feature_importances_'):
    importances = best_rf_classifier.feature_importances_
    sorted_idx = np.argsort(importances)[::-1]
    sorted_features = np.array(feature_cols)[sorted_idx]

    plt.figure(figsize=(12, 8))
    sns.barplot(x=importances[sorted_idx], y=sorted_features, palette='crest')
    plt.title("Feature Importance (from Best Random Forest Model - Adjusted Class Weights - 1.2x Boost)")
    plt.tight_layout()
    plt.show()
else:
    print("Cannot retrieve feature importances from the best classifier. It might not be a tree-based model.")


# --- [12] Learning Curve (จาก Best Model) ---
train_sizes, train_scores, test_scores = learning_curve(
    final_pipeline, X_train, y_train, cv=5, scoring=make_scorer(f1_score, average='macro'),
    train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1
)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel("Training Size")
plt.ylabel(f"F1-macro Score")
plt.title("Learning Curve (Best Random Forest Model - Adjusted Class Weights - 1.2x Boost)")
plt.legend()
plt.tight_layout()
plt.show()
```
