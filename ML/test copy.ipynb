```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score, StratifiedKFold


# --- 1. โหลดข้อมูล ---
df = pd.read_csv('food_menu.csv')

# --- 2. ปรับสมดุล class โดยการ Random oversampling class ที่มีน้อย ---
class_counts = df['food_type'].value_counts()
max_count = class_counts.max()
df_balanced = pd.DataFrame()

for food_class, count in class_counts.items():
    class_df = df[df['food_type'] == food_class]
    if count < max_count:
        repeat_factor = int(np.ceil(max_count / count))
        class_df_duplicated = pd.concat([class_df] * repeat_factor, ignore_index=True)
        class_df_balanced = class_df_duplicated.sample(n=max_count, random_state=42)
    else:
        class_df_balanced = class_df
    df_balanced = pd.concat([df_balanced, class_df_balanced], ignore_index=True)

df = df_balanced.copy()

# --- 3. Feature Engineering ---
category_encoder = LabelEncoder()
df['category_encoded'] = category_encoder.fit_transform(df['category'])
df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)

# --- 4. เลือก Features และ Target ---
X = df[['calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium',
        'category_encoded', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio']]
y = df['food_type']

# --- 5. Encode target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# แสดง mapping ระหว่าง label เดิม กับ ตัวเลขที่แปลง
for idx, class_label in enumerate(le.classes_):
    print(f"{class_label} --> {idx}")

# --- 6. Train/Test Split ก่อน ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)

# --- 7. จัดการ missing values เฉพาะบน train ---
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
# 
X_test_imputed = imputer.transform(X_test)

# --- 8. Scaling เฉพาะบน train ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)


print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")

# --- 9. สร้างและฝึกโมเดล (ปรับปรุง) ---
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_split=5, min_samples_leaf=2, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=10, weights='distance', metric='euclidean'),
    "Decision Tree": DecisionTreeClassifier(random_state=42, class_weight='balanced')
}

trained_models = {} # Dictionary ใหม่สำหรับเก็บโมเดลที่ฝึกแล้ว

print("\n--- Training Models ---")
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train_scaled, y_train)
    trained_models[name] = model # เก็บโมเดลที่ฝึกแล้ว

# --- 10.1 ประเมินผลบน Train Set (ปรับปรุง) ---
train_accuracies = []
class_labels = ['Low Sugar', 'Balanced', 'Low Sodium', 'Low Fat'] # ใช้ le.classes_ เพื่อความถูกต้องและยืดหยุ่น
print("\n--- Evaluating Models on Train Set ---")
for name, model in trained_models.items(): # ใช้ trained_models
    y_train_pred = model.predict(X_train_scaled)
    acc = accuracy_score(y_train, y_train_pred)
    report = classification_report(y_train, y_train_pred, target_names=class_labels)

    train_accuracies.append(acc)

    print(f"\n--- {name} Train Results ---")
    print(f"Accuracy: {acc:.2f}")
    print("Classification Report:")
    print(report)

    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, display_labels=class_labels,
                                             cmap='Blues', ax=ax)
    ax.set_title(f"Confusion Matrix: {name} (Train Set)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- 10.2 ประเมินผลบน Test Set (ปรับปรุง) ---
test_accuracies = []
model_names = [] # จะเป็นชื่อโมเดลเดิมจาก models.keys()
# class_labels_en = ['Low Sugar', 'Balanced', 'Low Sodium', 'Low Fat'] # ไม่จำเป็น ต้องใช้ le.classes_
class_labels_en = ['Low Sugar', 'Balanced', 'Low Sodium', 'Low Fat'] # ใช้ le.classes_ เหมือนเดิม

print("\n--- Evaluating Models on Test Set ---")
for name, model in trained_models.items(): # ใช้ trained_models
    y_test_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_test_pred)
    report = classification_report(y_test, y_test_pred, target_names=class_labels_en)

    test_accuracies.append(acc)
    model_names.append(name)

    print(f"\n--- {name} Test Results ---")
    print(f"Accuracy: {acc:.2f}")
    print("Classification Report:")
    print(report)

    fig, ax = plt.subplots(figsize=(8, 6))
    ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels_en,
                                             cmap='Blues', ax=ax)
    ax.set_title(f"Confusion Matrix: {name} (Test Set)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- 10.3 Cross Validation (ใหม่) ---
cv_scores = {}
cv_stddev = {}
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- Cross Validation Scores ---")
for name, model in models.items():  # ใช้ models แทน trained_models เพื่อไม่ overfit จาก train/test split
    scores = cross_val_score(model, X, y_encoded, cv=kfold, scoring='accuracy')
    cv_scores[name] = scores.mean()
    cv_stddev[name] = scores.std()
    
    print(f"{name}: Mean Accuracy = {scores.mean():.3f}, Std = {scores.std():.3f}")


# --- 11.1 กราฟเปรียบเทียบ Accuracy Train ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=train_accuracies, y=list(models.keys()), palette="viridis", orient='h')
for i, v in enumerate(train_accuracies):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold') 
plt.title('Model Accuracy Comparison on Train Set', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.savefig('model_comparison_train.png', dpi=300)
plt.show()

# --- 11.2 กราฟเปรียบเทียบ Accuracy Test ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=test_accuracies, y=model_names, palette="magma", orient='h')
for i, v in enumerate(test_accuracies):
    ax.text(v + 0.01, i, f"{v:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Accuracy Comparison on Test Set', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.savefig('model_comparison_test.png', dpi=300)
plt.show()

# --- 11.3 กราฟเปรียบเทียบ Cross Validation Accuracy ---
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=list(cv_scores.values()), y=list(cv_scores.keys()), palette="coolwarm", orient='h')
for i, (mean_score, std) in enumerate(zip(cv_scores.values(), cv_stddev.values())):
    ax.text(mean_score + 0.01, i, f"{mean_score:.2f} ± {std:.2f}", color='black', va='center', fontweight='bold')
plt.title('Model Cross-Validated Accuracy (5-Fold)', fontsize=16)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.xlim(0, 1.0)
plt.tight_layout()
plt.savefig('model_comparison_cv.png', dpi=300)
plt.show()



```
