```python
# --- [1] Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, learning_curve
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

# --- [2] Load Dataset ---
df = pd.read_csv('food_menu.csv')

# --- [3] Feature Engineering ---
# Encode category
df['category_encoded'] = LabelEncoder().fit_transform(df['category'])

# Extract numeric from 'amount'
df['amount_numeric'] = df['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)
df['sodium_per_calorie'] = df['sodium'] / (df['calories'] + 1e-8)

# Density & per-calorie features
df['calorie_density'] = df['calories'] / (df['amount_numeric'] + 1e-8)
df['sugar_per_calorie'] = df['sugar'] / (df['calories'] + 1e-8)
df['fat_per_calorie'] = df['fat'] / (df['calories'] + 1e-8)
df['protein_per_calorie'] = df['protein'] / (df['calories'] + 1e-8)
df['carb_per_calorie'] = df['carbohydrate'] / (df['calories'] + 1e-8)
df['protein_to_fat_ratio'] = df['protein'] / (df['fat'] + 1e-8)

# Density by amount
df['protein_density'] = df['protein'] / (df['amount_numeric'] + 1e-8)
df['fat_density'] = df['fat'] / (df['amount_numeric'] + 1e-8)
df['carb_density'] = df['carbohydrate'] / (df['amount_numeric'] + 1e-8)
df['sugar_density'] = df['sugar'] / (df['amount_numeric'] + 1e-8)
df['sodium_density'] = df['sodium'] / (df['amount_numeric'] + 1e-8)

# Extra ratios and densities
df['fiber_per_calorie'] = df['fiber'] / (df['calories'] + 1e-8)
df['fiber_to_carb_ratio'] = df['fiber'] / (df['carbohydrate'] + 1e-8)
df['sodium_to_sugar_ratio'] = df['sodium'] / (df['sugar'] + 1e-8)
df['saturated_fat_per_calorie'] = df['saturated_fat'] / (df['calories'] + 1e-8)
df['saturated_fat_to_total_fat_ratio'] = df['saturated_fat'] / (df['fat'] + 1e-8)
df['fiber_density'] = df['fiber'] / (df['amount_numeric'] + 1e-8)
df['saturated_fat_density'] = df['saturated_fat'] / (df['amount_numeric'] + 1e-8)

# --- Feature Engineering à¹€à¸žà¸´à¹ˆà¸¡à¹€à¸•à¸´à¸¡à¸ªà¸³à¸«à¸£à¸±à¸šà¸„à¸¥à¸²à¸ª 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³' à¹à¸¥à¸° 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³' ---

# 1. à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¹à¸šà¸š Interaction (à¸à¸²à¸£à¸£à¸§à¸¡à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´)
# à¸à¸²à¸£à¸„à¸¹à¸“à¸«à¸£à¸·à¸­à¸«à¸²à¸£à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¸—à¸µà¹ˆà¹€à¸à¸µà¹ˆà¸¢à¸§à¸‚à¹‰à¸­à¸‡ à¸­à¸²à¸ˆà¸Šà¹ˆà¸§à¸¢à¸‚à¸¢à¸²à¸¢à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡
df['sugar_x_fat'] = df['sugar'] * df['fat']
df['sugar_plus_carb'] = df['sugar'] + df['carbohydrate'] # à¸ªà¸±à¸”à¸ªà¹ˆà¸§à¸™à¸£à¸§à¸¡à¸‚à¸­à¸‡à¸žà¸¥à¸±à¸‡à¸‡à¸²à¸™à¸ˆà¸²à¸à¸™à¹‰à¸³à¸•à¸²à¸¥à¹à¸¥à¸°à¸„à¸²à¸£à¹Œà¹‚à¸šà¹„à¸®à¹€à¸”à¸£à¸•
df['fat_plus_saturated_fat'] = df['fat'] + df['saturated_fat'] # à¹„à¸‚à¸¡à¸±à¸™à¸£à¸§à¸¡à¸—à¸µà¹ˆà¸ªà¸³à¸„à¸±à¸

# 2. à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¹à¸šà¸š Polynomial (à¸à¸²à¸£à¸¢à¸à¸à¸³à¸¥à¸±à¸‡ - à¸­à¸²à¸ˆà¸Šà¹ˆà¸§à¸¢à¹ƒà¸«à¹‰à¹€à¸ªà¹‰à¸™à¹à¸šà¹ˆà¸‡à¸„à¸¥à¸²à¸ªà¹„à¸¡à¹ˆà¹€à¸›à¹‡à¸™à¹€à¸Šà¸´à¸‡à¹€à¸ªà¹‰à¸™)
# df['sugar_sq'] = df['sugar']**2 # à¸–à¹‰à¸²à¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¹€à¸žà¸´à¹ˆà¸¡à¸„à¸§à¸²à¸¡à¹‚à¸”à¸”à¹€à¸”à¹ˆà¸™à¸‚à¸­à¸‡à¸„à¹ˆà¸²à¸™à¹‰à¸³à¸•à¸²à¸¥à¸—à¸µà¹ˆà¸ªà¸¹à¸‡
# df['fat_sq'] = df['fat']**2 # à¸–à¹‰à¸²à¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¹€à¸žà¸´à¹ˆà¸¡à¸„à¸§à¸²à¸¡à¹‚à¸”à¸”à¹€à¸”à¹ˆà¸™à¸‚à¸­à¸‡à¸„à¹ˆà¸²à¹„à¸‚à¸¡à¸±à¸™à¸—à¸µà¹ˆà¸ªà¸¹à¸‡
# df['sugar_sqrt'] = np.sqrt(df['sugar']) # à¸–à¹‰à¸²à¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸¥à¸”à¸œà¸¥à¸à¸£à¸°à¸—à¸šà¸‚à¸­à¸‡à¸„à¹ˆà¸²à¸ªà¸¹à¸‡à¹† à¹à¸¥à¸°à¸”à¸¶à¸‡à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡à¸‚à¸­à¸‡à¸„à¹ˆà¸²à¸•à¹ˆà¸³à¹† (à¸ªà¸³à¸«à¸£à¸±à¸šà¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³)
# df['fat_sqrt'] = np.sqrt(df['fat']) # à¸„à¸¥à¹‰à¸²à¸¢à¸à¸±à¸™à¸ªà¸³à¸«à¸£à¸±à¸šà¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³
# *à¸‚à¹‰à¸­à¸„à¸§à¸£à¸£à¸°à¸§à¸±à¸‡: Polynomial Features à¸­à¸²à¸ˆà¸—à¸³à¹ƒà¸«à¹‰à¹€à¸à¸´à¸” Multicollinearity à¹à¸¥à¸° Overfitting à¹„à¸”à¹‰à¸‡à¹ˆà¸²à¸¢ à¸•à¹‰à¸­à¸‡à¸£à¸°à¸§à¸±à¸‡*

# 3. à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¹€à¸Šà¸´à¸‡ Threshold (à¸­à¹‰à¸²à¸‡à¸­à¸´à¸‡à¸ˆà¸²à¸à¹€à¸à¸“à¸‘à¹Œà¸—à¸µà¹ˆà¸„à¸²à¸”à¸§à¹ˆà¸²à¹€à¸à¸µà¹ˆà¸¢à¸§à¸‚à¹‰à¸­à¸‡à¸à¸±à¸š 'à¸•à¹ˆà¸³')
# à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸à¹€à¸£à¸²à¹„à¸¡à¹ˆà¸¡à¸µà¹€à¸à¸“à¸‘à¹Œà¸—à¸µà¹ˆà¸Šà¸±à¸”à¹€à¸ˆà¸™à¸§à¹ˆà¸² "à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³" à¸«à¸£à¸·à¸­ "à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³" à¸„à¸·à¸­à¹€à¸—à¹ˆà¸²à¹„à¸«à¸£à¹ˆ
# à¹€à¸£à¸²à¸ªà¸²à¸¡à¸²à¸£à¸–à¹ƒà¸Šà¹‰à¸„à¹ˆà¸² Median à¸«à¸£à¸·à¸­ Quantile à¸‚à¸­à¸‡ 'sugar'/'fat' à¸ˆà¸²à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸” à¸«à¸£à¸·à¸­à¸ˆà¸²à¸à¸„à¸¥à¸²à¸ªà¸—à¸µà¹ˆà¸¡à¸µà¸›à¸±à¸à¸«à¸²
# à¸¡à¸²à¸ªà¸£à¹‰à¸²à¸‡à¹€à¸›à¹‡à¸™à¹€à¸à¸“à¸‘à¹Œà¹€à¸šà¸·à¹‰à¸­à¸‡à¸•à¹‰à¸™

# à¸„à¸³à¸™à¸§à¸“ Median à¸«à¸£à¸·à¸­ Quantile à¸ˆà¸²à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
median_sugar = df['sugar'].median()
median_fat = df['fat'].median()
q25_sugar = df['sugar'].quantile(0.25) # 25th percentile
q25_fat = df['fat'].quantile(0.25) # 25th percentile

# à¸ªà¸£à¹‰à¸²à¸‡à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´ Binary (0 à¸«à¸£à¸·à¸­ 1) à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰à¹€à¸à¸“à¸‘à¹Œ
# is_low_sugar: 1 à¸–à¹‰à¸² 'sugar' à¸•à¹ˆà¸³à¸à¸§à¹ˆà¸²à¸„à¹ˆà¸² Median/Quantile
df['is_low_sugar_median'] = (df['sugar'] < median_sugar).astype(int)
df['is_low_sugar_q25'] = (df['sugar'] < q25_sugar).astype(int)

# is_low_fat: 1 à¸–à¹‰à¸² 'fat' à¸•à¹ˆà¸³à¸à¸§à¹ˆà¸²à¸„à¹ˆà¸² Median/Quantile
df['is_low_fat_median'] = (df['fat'] < median_fat).astype(int)
df['is_low_fat_q25'] = (df['fat'] < q25_fat).astype(int)

# à¸­à¸²à¸ˆà¸¥à¸­à¸‡à¸ªà¸£à¹‰à¸²à¸‡à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¸£à¸§à¸¡à¸‚à¸­à¸‡à¸ªà¸²à¸£à¸­à¸²à¸«à¸²à¸£à¸­à¸·à¹ˆà¸™à¹† à¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¹€à¸à¸µà¹ˆà¸¢à¸§à¸‚à¹‰à¸­à¸‡ à¹€à¸žà¸·à¹ˆà¸­à¹à¸¢à¸à¸„à¸¥à¸²à¸ª "à¸ªà¸¡à¸”à¸¸à¸¥" à¸­à¸­à¸à¹„à¸›
df['total_macro_nutrients'] = df['protein'] + df['carbohydrate'] + df['fat']

# à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¸—à¸µà¹ˆà¹€à¸™à¹‰à¸™à¸„à¸§à¸²à¸¡à¹à¸•à¸à¸•à¹ˆà¸²à¸‡à¸ˆà¸²à¸à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢/Median à¸‚à¸­à¸‡à¸„à¸¥à¸²à¸ªà¸­à¸·à¹ˆà¸™
# à¸ˆà¸³à¹€à¸›à¹‡à¸™à¸•à¹‰à¸­à¸‡à¸¡à¸µ `food_type` à¹€à¸žà¸·à¹ˆà¸­à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡à¸„à¸¥à¸²à¸ªà¸­à¸·à¹ˆà¸™à¹†
# à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸à¹€à¸£à¸²à¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¹à¸¢à¸ 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³' à¹à¸¥à¸° 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³'
# à¹ƒà¸«à¹‰à¸¥à¸­à¸‡à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡à¸™à¹‰à¸³à¸•à¸²à¸¥à¹à¸¥à¸°à¹„à¸‚à¸¡à¸±à¸™ **à¸ˆà¸²à¸à¸„à¸¥à¸²à¸ªà¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¹ƒà¸Šà¹ˆà¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³/à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³**
# à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸Šà¹‰à¹€à¸›à¹‡à¸™à¸à¸²à¸™à¹ƒà¸™à¸à¸²à¸£à¹€à¸›à¸£à¸µà¸¢à¸šà¹€à¸—à¸µà¸¢à¸š

# à¸‚à¸±à¹‰à¸™à¹à¸£à¸: à¸à¸³à¸«à¸™à¸” Class Labels (à¸–à¹‰à¸²à¸¢à¸±à¸‡à¹„à¸¡à¹ˆà¸¡à¸µ)
# à¸•à¹‰à¸­à¸‡à¹à¸™à¹ˆà¹ƒà¸ˆà¸§à¹ˆà¸²à¹„à¸”à¹‰à¸£à¸±à¸™à¸ªà¹ˆà¸§à¸™ LabelEncoder(y) à¸¡à¸²à¹à¸¥à¹‰à¸§ à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¸£à¸¹à¹‰à¸§à¹ˆà¸² 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³' à¹à¸¥à¸° 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³' à¸„à¸·à¸­ index à¸­à¸°à¹„à¸£
# à¸ªà¸¡à¸¡à¸•à¸´à¸§à¹ˆà¸² 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³' à¸„à¸·à¸­ index 0, 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³' à¸„à¸·à¸­ index 3 (à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸ˆà¸²à¸ print("Class Mapping:") à¸à¹ˆà¸­à¸™à¸«à¸™à¹‰à¸²à¸™à¸µà¹‰)
# à¸•à¹‰à¸­à¸‡à¹ƒà¸Šà¹‰ y_encoded à¹ƒà¸™à¸à¸²à¸£à¸„à¸³à¸™à¸§à¸“
# à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸à¹‚à¸„à¹‰à¸”à¸™à¸µà¹‰à¸£à¸±à¸™à¸à¹ˆà¸­à¸™ y_encoded à¸ˆà¸°à¸–à¸¹à¸à¸ªà¸£à¹‰à¸²à¸‡, à¸•à¹‰à¸­à¸‡à¸—à¸³à¸à¸²à¸£ map y à¸à¸¥à¸±à¸šà¹„à¸›à¸à¹ˆà¸­à¸™ à¸«à¸£à¸·à¸­à¸„à¸³à¸™à¸§à¸“à¸«à¸¥à¸±à¸‡à¸ˆà¸²à¸ y_encoded à¸–à¸¹à¸à¸ªà¸£à¹‰à¸²à¸‡
# à¸ªà¸³à¸«à¸£à¸±à¸šà¸•à¸­à¸™à¸™à¸µà¹‰ à¸‚à¸­à¸ªà¸¡à¸¡à¸•à¸´à¸§à¹ˆà¸² 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³' à¹à¸¥à¸° 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³' à¹€à¸›à¹‡à¸™à¸„à¹ˆà¸²à¹ƒà¸™à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'food_type' à¹‚à¸”à¸¢à¸•à¸£à¸‡
# à¸–à¹‰à¸²à¸„à¸¸à¸“à¸¡à¸µ `food_type` à¹€à¸›à¹‡à¸™ string à¹ƒà¸™ df à¹ƒà¸«à¹‰à¹ƒà¸Šà¹‰à¹à¸šà¸šà¸™à¸µà¹‰
# avg_sugar_not_low_sugar = df[df['food_type'] != 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³']['sugar'].mean()
# avg_fat_not_low_fat = df[df['food_type'] != 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³']['fat'].mean()

# à¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¹‚à¸„à¹‰à¸” Feature Engineering à¸­à¸¢à¸¹à¹ˆà¹ƒà¸™à¸šà¸¥à¹‡à¸­à¸à¹€à¸”à¸µà¸¢à¸§à¸à¸±à¸™à¸à¸±à¸šà¹‚à¸„à¹‰à¸”à¸—à¸µà¹ˆà¸„à¸¸à¸“à¹ƒà¸«à¹‰à¸¡à¸²
# à¹€à¸£à¸²à¸ˆà¸°à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¹€à¸«à¸¥à¹ˆà¸²à¸™à¸µà¹‰à¸«à¸¥à¸±à¸‡à¸ˆà¸²à¸à¸—à¸µà¹ˆ `food_type` (original string) à¸¡à¸µà¸­à¸¢à¸¹à¹ˆ
# à¹‚à¸”à¸¢à¸­à¸²à¸ˆà¸ˆà¸°à¸•à¹‰à¸­à¸‡à¸£à¸±à¸™à¹‚à¸„à¹‰à¸”à¸ªà¹ˆà¸§à¸™ [2] à¹à¸¥à¸° [3] à¹ƒà¸«à¸¡à¹ˆà¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
# à¸«à¸£à¸·à¸­à¹ƒà¸«à¹‰à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹€à¸«à¸¥à¹ˆà¸²à¸™à¸µà¹‰à¸¡à¸²à¹€à¸žà¸´à¹ˆà¸¡
# à¹à¸•à¹ˆà¸–à¹‰à¸²à¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸ªà¸£à¹‰à¸²à¸‡ FE à¸—à¸µà¹ˆà¸‹à¸±à¸šà¸‹à¹‰à¸­à¸™à¸‚à¸¶à¹‰à¸™ à¸„à¸§à¸£à¸—à¸³à¹ƒà¸™à¸‚à¸±à¹‰à¸™à¸•à¸­à¸™à¹à¸¢à¸à¸•à¹ˆà¸²à¸‡à¸«à¸²à¸ à¸«à¸£à¸·à¸­à¸«à¸¥à¸±à¸‡ y_encoded à¸–à¸¹à¸à¸ªà¸£à¹‰à¸²à¸‡
# **à¸ªà¸³à¸«à¸£à¸±à¸šà¸•à¸­à¸™à¸™à¸µà¹‰ à¹€à¸£à¸²à¸ˆà¸°à¸‚à¹‰à¸²à¸¡à¸ªà¹ˆà¸§à¸™à¸™à¸µà¹‰à¹„à¸›à¸à¹ˆà¸­à¸™ à¸«à¸²à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Food_type à¸—à¸µà¹ˆà¹à¸—à¹‰à¸ˆà¸£à¸´à¸‡à¹„à¸¡à¹ˆà¹„à¸”à¹‰à¸­à¸¢à¸¹à¹ˆà¹ƒà¸™ df à¸™à¸µà¹‰à¹‚à¸”à¸¢à¸•à¸£à¸‡**
# (à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸à¹ƒà¸™à¸ªà¹ˆà¸§à¸™ [4] à¸–à¸¶à¸‡à¸ˆà¸°à¸¡à¸µà¸à¸²à¸£à¸à¸³à¸«à¸™à¸” y = df['food_type'] à¹à¸¥à¸° Encode)
# à¸”à¸±à¸‡à¸™à¸±à¹‰à¸™ à¹€à¸£à¸²à¸ˆà¸°à¹€à¸™à¹‰à¸™à¹„à¸›à¸—à¸µà¹ˆà¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¸—à¸µà¹ˆà¸ªà¸£à¹‰à¸²à¸‡à¹„à¸”à¹‰à¸ˆà¸²à¸à¸„à¹ˆà¸²à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¹€à¸—à¹ˆà¸²à¸™à¸±à¹‰à¸™à¸à¹ˆà¸­à¸™

# --- [4] Select Features & Target ---
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie', 'saturated_fat_to_total_fat_ratio',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density', 'category_encoded',
    # à¹€à¸žà¸´à¹ˆà¸¡à¸„à¸¸à¸“à¸ªà¸¡à¸šà¸±à¸•à¸´à¹ƒà¸«à¸¡à¹ˆà¸—à¸µà¹ˆà¸ªà¸£à¹‰à¸²à¸‡à¸‚à¸¶à¹‰à¸™
    'sugar_x_fat',
    'sugar_plus_carb',
    'fat_plus_saturated_fat',
    'is_low_sugar_median',
    'is_low_sugar_q25',
    'is_low_fat_median',
    'is_low_fat_q25',
    'total_macro_nutrients'
    # à¸–à¹‰à¸²à¸¡à¸µà¸à¸²à¸£à¹€à¸žà¸´à¹ˆà¸¡ Polynomial Features à¸•à¹‰à¸­à¸‡à¹€à¸žà¸´à¹ˆà¸¡à¸•à¸£à¸‡à¸™à¸µà¹‰à¸”à¹‰à¸§à¸¢ à¹€à¸Šà¹ˆà¸™ 'sugar_sq', 'fat_sq'
]
X = df[feature_cols]
y = df['food_type']

print("--- New Features Added for targeted separation: ---")
print("sugar_x_fat: Product of sugar and fat")
print("sugar_plus_carb: Sum of sugar and carbohydrate")
print("fat_plus_saturated_fat: Sum of fat and saturated fat")
print("is_low_sugar_median: Binary flag if sugar is below dataset median")
print("is_low_sugar_q25: Binary flag if sugar is below dataset 25th percentile")
print("is_low_fat_median: Binary flag if fat is below dataset median")
print("is_low_fat_q25: Binary flag if fat is below dataset 25th percentile")
print("total_macro_nutrients: Sum of protein, carbohydrate, and fat")
print(f"\nTotal features now: {len(feature_cols)}")

# --- [5] Encode Target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_labels = le.classes_

print("Class Mapping:")
for idx, name in enumerate(class_labels):
    print(f"{name} --> {idx}")

# --- [6] Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

# --- [7] Build Pipeline ---
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(
        n_estimators=50,
        max_depth=10,
        min_samples_split=2,
        min_samples_leaf=2,
        random_state=42,
        class_weight='balanced'
    ))
])

# --- [8] Train Pipeline ---
pipeline.fit(X_train, y_train)

# --- [9] Evaluate Train Set ---
y_train_pred = pipeline.predict(X_train)
print("\nðŸ“Š Train Set Evaluation:")
print(f"Accuracy: {accuracy_score(y_train, y_train_pred):.2f}")
print(classification_report(y_train, y_train_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Train Set")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- [10] Evaluate Test Set ---
y_test_pred = pipeline.predict(X_test)
print("\nðŸ“Š Test Set Evaluation:")
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.2f}")
print(classification_report(y_test, y_test_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Test Set")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# --- [11] Feature Importance ---
# Bypass Pipeline (SMOTE affects train only)
X_train_imp = SimpleImputer(strategy='mean').fit_transform(X_train)
X_train_scaled = StandardScaler().fit_transform(X_train_imp)
X_bal, y_bal = SMOTE(random_state=42).fit_resample(X_train_scaled, y_train)

rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X_bal, y_bal)

importances = rf.feature_importances_
sorted_idx = np.argsort(importances)[::-1]
sorted_features = np.array(feature_cols)[sorted_idx]

plt.figure(figsize=(10, 6))
sns.barplot(x=importances[sorted_idx], y=sorted_features, palette='crest')
plt.title("Feature Importance")
plt.tight_layout()
plt.show()

# --- [12] Learning Curve ---
train_sizes, train_scores, test_scores = learning_curve(
    rf, X_bal, y_bal, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1
)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve")
plt.legend()
plt.tight_layout()
plt.show()

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for label in np.unique(y):
    plt.scatter(X_pca[y == label, 0], X_pca[y == label, 1], label=label, alpha=0.6)
plt.title("PCA: Class Separation")
plt.legend()
plt.show()


from sklearn.tree import plot_tree

# à¹€à¸¥à¸·à¸­à¸à¸•à¹‰à¸™à¹„à¸¡à¹‰à¸•à¹‰à¸™à¸—à¸µà¹ˆ 0 à¸ˆà¸²à¸ Random Forest à¸—à¸µà¹ˆ train à¹„à¸›à¹à¸¥à¹‰à¸§
tree_to_plot = rf.estimators_[0]  # à¹€à¸¥à¸·à¸­à¸à¸•à¹‰à¸™à¹„à¸¡à¹‰à¸•à¹‰à¸™à¸—à¸µà¹ˆ 1 (index 0)

# à¸ªà¸£à¹‰à¸²à¸‡à¹à¸œà¸™à¸œà¸±à¸‡
plt.figure(figsize=(24, 12))
plot_tree(tree_to_plot,
          feature_names=feature_cols,
          class_names=class_labels,
          filled=True,
          rounded=True,
          fontsize=8,
          max_depth=3  # à¸›à¸£à¸±à¸šà¹€à¸žà¸·à¹ˆà¸­à¹ƒà¸«à¹‰à¹à¸ªà¸”à¸‡à¹à¸„à¹ˆà¸£à¸°à¸”à¸±à¸šà¸šà¸™à¹†
         )

plt.title("ðŸŒ³ Random Forest: Tree Visualization (Tree #0, max_depth=3)")
plt.tight_layout()
plt.show()


```

```python
# --- [1] Import Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, learning_curve, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay, make_scorer, f1_score
from sklearn.decomposition import PCA
from sklearn.tree import plot_tree

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.utils import class_weight

# --- [2] Load Dataset ---
df = pd.read_csv('food_menu.csv')

# --- [3] Feature Engineering (à¹ƒà¸Šà¹‰à¸Šà¸¸à¸”à¸¥à¹ˆà¸²à¸ªà¸¸à¸”à¸—à¸µà¹ˆà¸¥à¸š 'saturated_fat_to_total_fat_ratio' à¹à¸¥à¹‰à¸§) ---
# Encode category
df['category_encoded'] = LabelEncoder().fit_transform(df['category'])

# Extract numeric from 'amount'
df['amount_numeric'] = df['amount'].str.extract(r'(\d+\.?\d*)').astype(float)

# Base ratios
df['sugar_to_carb_ratio'] = df['sugar'] / (df['carbohydrate'] + 1e-8)
df['fat_to_calorie_ratio'] = df['fat'] / (df['calories'] + 1e-8)
df['sodium_per_calorie'] = df['sodium'] / (df['calories'] + 1e-8)

# Density & per-calorie features
df['calorie_density'] = df['calories'] / (df['amount_numeric'] + 1e-8)
df['sugar_per_calorie'] = df['sugar'] / (df['calories'] + 1e-8)
df['fat_per_calorie'] = df['fat'] / (df['calories'] + 1e-8)
df['protein_per_calorie'] = df['protein'] / (df['calories'] + 1e-8)
df['carb_per_calorie'] = df['carbohydrate'] / (df['calories'] + 1e-8)
df['protein_to_fat_ratio'] = df['protein'] / (df['fat'] + 1e-8)

# Density by amount
df['protein_density'] = df['protein'] / (df['amount_numeric'] + 1e-8)
df['fat_density'] = df['fat'] / (df['amount_numeric'] + 1e-8)
df['carb_density'] = df['carbohydrate'] / (df['amount_numeric'] + 1e-8)
df['sugar_density'] = df['sugar'] / (df['amount_numeric'] + 1e-8)
df['sodium_density'] = df['sodium'] / (df['amount_numeric'] + 1e-8)

# Extra ratios and densities
df['fiber_per_calorie'] = df['fiber'] / (df['calories'] + 1e-8)
df['fiber_to_carb_ratio'] = df['fiber'] / (df['carbohydrate'] + 1e-8)
df['sodium_to_sugar_ratio'] = df['sodium'] / (df['sugar'] + 1e-8)
df['saturated_fat_per_calorie'] = df['saturated_fat'] / (df['calories'] + 1e-8)
df['fiber_density'] = df['fiber'] / (df['amount_numeric'] + 1e-8)
df['saturated_fat_density'] = df['saturated_fat'] / (df['amount_numeric'] + 1e-8)

# Interaction Features
df['sugar_x_fat'] = df['sugar'] * df['fat']
df['sugar_plus_carb'] = df['sugar'] + df['carbohydrate']
df['fat_plus_saturated_fat'] = df['fat'] + df['saturated_fat']
df['total_macro_nutrients'] = df['protein'] + df['carbohydrate'] + df['fat']

# Binary Features (à¸„à¸‡à¹„à¸§à¹‰à¸•à¸²à¸¡à¹€à¸”à¸´à¸¡)
median_sugar = df['sugar'].median()
median_fat = df['fat'].median()
q25_sugar = df['sugar'].quantile(0.25)
q25_fat = df['fat'].quantile(0.25)

df['is_low_sugar_median'] = (df['sugar'] < median_sugar).astype(int)
df['is_low_sugar_q25'] = (df['sugar'] < q25_sugar).astype(int)
df['is_low_fat_median'] = (df['fat'] < median_fat).astype(int)
df['is_low_fat_q25'] = (df['fat'] < q25_fat).astype(int)


# RDI-based Features
RDI_CALORIES = 2000
RDI_SUGAR = 25
RDI_FAT = 65
RDI_SODIUM = 2000
RDI_PROTEIN = 50
RDI_CARB = 275
RDI_FIBER = 28
RDI_SAT_FAT = 20

df['sugar_rdi_ratio'] = df['sugar'] / RDI_SUGAR
df['fat_rdi_ratio'] = df['fat'] / RDI_FAT
df['sodium_rdi_ratio'] = df['sodium'] / RDI_SODIUM
df['protein_rdi_ratio'] = df['protein'] / RDI_PROTEIN
df['carb_rdi_ratio'] = df['carbohydrate'] / RDI_CARB
df['fiber_rdi_ratio'] = df['fiber'] / RDI_FIBER
df['sat_fat_rdi_ratio'] = df['saturated_fat'] / RDI_SAT_FAT
df['calories_rdi_ratio'] = df['calories'] / RDI_CALORIES

df['is_high_sugar_rdi'] = (df['sugar_rdi_ratio'] > (1/3)).astype(int)
df['is_low_sugar_rdi'] = (df['sugar_rdi_ratio'] < (1/25)).astype(int)

df['is_high_fat_rdi'] = (df['fat_rdi_ratio'] > (1/3)).astype(int)
df['is_low_fat_rdi'] = (df['fat_rdi_ratio'] < (1/25)).astype(int)

df['is_high_sodium_rdi'] = (df['sodium_rdi_ratio'] > (1/3)).astype(int)
df['is_low_sodium_rdi'] = (df['sodium_rdi_ratio'] < (1/25)).astype(int)

df['is_high_sat_fat_rdi'] = (df['sat_fat_rdi_ratio'] > (1/3)).astype(int)

# Specific Threshold Features (à¸„à¸‡à¹„à¸§à¹‰)
df['is_very_low_sugar'] = (df['sugar'] < 5).astype(int)
df['is_zero_sugar'] = (df['sugar'] == 0).astype(int)

df['is_very_low_fat'] = (df['fat'] < 3).astype(int)
df['is_fat_free'] = (df['fat'] == 0).astype(int)

df['is_fda_low_sodium'] = (df['sodium'] <= 140).astype(int)


# --- [4] Select Features & Target ---
feature_cols = [
    'calories', 'protein', 'carbohydrate', 'sugar', 'fat', 'sodium', 'fiber',
    'saturated_fat', 'sugar_to_carb_ratio', 'fat_to_calorie_ratio', 'sodium_per_calorie',
    'calorie_density', 'sugar_per_calorie', 'fat_per_calorie', 'protein_per_calorie',
    'carb_per_calorie', 'protein_to_fat_ratio', 'fiber_per_calorie', 'fiber_to_carb_ratio',
    'sodium_to_sugar_ratio', 'saturated_fat_per_calorie',
    'protein_density', 'fat_density', 'carb_density', 'sugar_density', 'sodium_density',
    'fiber_density', 'saturated_fat_density', 'category_encoded',
    'sugar_x_fat',
    'sugar_plus_carb',
    'fat_plus_saturated_fat',
    'is_low_sugar_median',
    'is_low_sugar_q25',
    'is_low_fat_median',
    'is_low_fat_q25',
    'total_macro_nutrients',
    # RDI-based Features
    'sugar_rdi_ratio', 'fat_rdi_ratio', 'sodium_rdi_ratio', 'protein_rdi_ratio',
    'carb_rdi_ratio', 'fiber_rdi_ratio', 'sat_fat_rdi_ratio', 'calories_rdi_ratio',
    'is_high_sugar_rdi', 'is_low_sugar_rdi', 'is_high_fat_rdi', 'is_low_fat_rdi',
    'is_high_sodium_rdi', 'is_low_sodium_rdi', 'is_high_sat_fat_rdi',
    # Specific Threshold Features
    'is_very_low_sugar', 'is_zero_sugar', 'is_very_low_fat', 'is_fat_free', 'is_fda_low_sodium'
]
X = df[feature_cols]
y = df['food_type']

# --- [5] Encode Target ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_labels = le.classes_

print("Class Mapping:")
for idx, name in enumerate(class_labels):
    print(f"{name} --> {idx}")

# --- [6] Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

# --- [7] Build Pipeline (à¸ªà¸³à¸«à¸£à¸±à¸š GridSearchCV) ---
classes = np.unique(y_train)
weights = class_weight.compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights_dict = dict(zip(classes, weights))

print("\nCalculated Class Weights (initial balanced):")
for cls_idx, weight in class_weights_dict.items():
    print(f"Class {le.inverse_transform([cls_idx])[0]} (Index {cls_idx}): {weight:.2f}")

# **à¸›à¸£à¸±à¸š Class Weights à¸”à¹‰à¸§à¸¢à¸¡à¸·à¸­à¸ªà¸³à¸«à¸£à¸±à¸š 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³' à¹à¸¥à¸° 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³' (à¸à¸¥à¸±à¸šà¹„à¸›à¹ƒà¸Šà¹‰ 1.2x)**
# à¸”à¸¶à¸‡ index à¸‚à¸­à¸‡à¸„à¸¥à¸²à¸ª 'à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³' à¹à¸¥à¸° 'à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³'
fat_low_idx = le.transform(['à¹„à¸‚à¸¡à¸±à¸™à¸•à¹ˆà¸³'])[0]
sugar_low_idx = le.transform(['à¸™à¹‰à¸³à¸•à¸²à¸¥à¸•à¹ˆà¸³'])[0]

# à¸à¸¥à¸±à¸šà¹„à¸›à¹ƒà¸Šà¹‰ 1.2 à¹€à¸—à¹ˆà¸²à¸‚à¸­à¸‡à¸™à¹‰à¸³à¸«à¸™à¸±à¸à¸—à¸µà¹ˆà¸„à¸³à¸™à¸§à¸“à¹„à¸”à¹‰à¸ˆà¸²à¸ 'balanced'
class_weights_dict[fat_low_idx] = class_weights_dict[fat_low_idx] * 1.2
class_weights_dict[sugar_low_idx] = class_weights_dict[sugar_low_idx] * 1.2

print("\nAdjusted Class Weights (reverted to 1.2x boost):")
for cls_idx, weight in class_weights_dict.items():
    print(f"Class {le.inverse_transform([cls_idx])[0]} (Index {cls_idx}): {weight:.2f}")

base_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42, class_weight=class_weights_dict)) # à¹ƒà¸Šà¹‰ class_weight à¸—à¸µà¹ˆà¸›à¸£à¸±à¸šà¹à¸¥à¹‰à¸§
])

# --- [8] à¹ƒà¸Šà¹‰ Best Parameters à¸—à¸µà¹ˆà¹„à¸”à¹‰à¸ˆà¸²à¸à¸à¸²à¸£à¸£à¸±à¸™à¸à¹ˆà¸­à¸™à¸«à¸™à¹‰à¸²à¸™à¸µà¹‰ ---
print("\n--- Model Training with Adjusted Class Weights (Reverted to 1.2x Boost) ---")

# Best Parameters à¸ˆà¸²à¸à¸à¸²à¸£à¸£à¸±à¸™à¸„à¸£à¸±à¹‰à¸‡à¸¥à¹ˆà¸²à¸ªà¸¸à¸” (à¸—à¸µà¹ˆà¹„à¸”à¹‰ Accuracy 0.80)
best_params_for_classifier = {
    'criterion': 'gini',
    'max_depth': 15,
    'max_features': 0.7,
    'min_samples_leaf': 3,
    'min_samples_split': 10,
    'n_estimators': 350
}

final_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42, class_weight=class_weights_dict, **best_params_for_classifier))
])

final_pipeline.fit(X_train, y_train)

print("\n--- Model Training with Adjusted Class Weights (Reverted to 1.2x Boost) Completed ---")
print(f"Parameters used for classifier: {final_pipeline.named_steps['classifier'].get_params()}")


# --- [9] Evaluate Train Set (with Best Model) ---
y_train_pred = final_pipeline.predict(X_train)
print("\nðŸ“Š Train Set Evaluation (Best Model with Adjusted Class Weights - 1.2x Boost):")
print(f"Accuracy: {accuracy_score(y_train, y_train_pred):.2f}")
print(classification_report(y_train, y_train_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Train Set (Best Model with Adjusted Class Weights - 1.2x Boost)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- [10] Evaluate Test Set (with Best Model) ---
y_test_pred = final_pipeline.predict(X_test)
print("\nðŸ“Š Test Set Evaluation (Best Model with Adjusted Class Weights - 1.2x Boost):")
print(f"Accuracy: {accuracy_score(y_test, y_test_pred):.2f}")
print(classification_report(y_test, y_test_pred, target_names=class_labels))

plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, display_labels=class_labels, cmap='Blues')
plt.title("Confusion Matrix: Test Set (Best Model with Adjusted Class Weights - 1.2x Boost)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# --- [11] Feature Importance (à¸ˆà¸²à¸ Best Model) ---
best_rf_classifier = final_pipeline.named_steps['classifier']

if hasattr(best_rf_classifier, 'feature_importances_'):
    importances = best_rf_classifier.feature_importances_
    sorted_idx = np.argsort(importances)[::-1]
    sorted_features = np.array(feature_cols)[sorted_idx]

    plt.figure(figsize=(12, 8))
    sns.barplot(x=importances[sorted_idx], y=sorted_features, palette='crest')
    plt.title("Feature Importance (from Best Random Forest Model - Adjusted Class Weights - 1.2x Boost)")
    plt.tight_layout()
    plt.show()
else:
    print("Cannot retrieve feature importances from the best classifier. It might not be a tree-based model.")


# --- [12] Learning Curve (à¸ˆà¸²à¸ Best Model) ---
train_sizes, train_scores, test_scores = learning_curve(
    final_pipeline, X_train, y_train, cv=5, scoring=make_scorer(f1_score, average='macro'),
    train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1
)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
plt.xlabel("Training Size")
plt.ylabel(f"F1-macro Score")
plt.title("Learning Curve (Best Random Forest Model - Adjusted Class Weights - 1.2x Boost)")
plt.legend()
plt.tight_layout()
plt.show()
```
